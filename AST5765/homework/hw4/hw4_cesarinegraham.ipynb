```python
# Problem 2a. 
# Find a python fun. that draws a random sample of N draws from a Gaussian distribution \
# of width sigma and mean x

# using this in my stat class currently!
m = 10000 
mu = 55
sigma = 13
sample = np.random.normal(mu, sigma, size=m) #Draws m i.i.d samples from N(mu, sigma2). x is an array of length m. 

print(sample)
```

```python
# Problem 2b
# plot the histogram of your sample

plt.figure( figsize = (8, 6) )
plt.hist(sample, bins=np.arange(0, 101, 1), color='green', density=False, alpha=0.6, label="Sample Counts")
plt.title("Histogram of Gaussian Sample ($\mu=55$, $\sigma=13$, $N=10000$)")
plt.xlabel('x')
plt.ylabel('Counts')
plt.xlim(0, 100)
plt.grid(True)
plt.legend()
plt.savefig("prob2b_cesarinegraham_histogram.png")
plt.show()
```

```python
# Problem 2c
# overplot a guassian distribution with the same mean and sigma as your histogram

m = 10000 
mu = 55
sigma = 13
sample = np.random.normal(mu, sigma, size=m)

# plotting our sample or 'empirical' data
plt.hist(sample, bins=np.arange(0, 101, 1), color='green', density=False, alpha=0.6, label="Sample Counts")

# creating our curve
x = np.linspace(0, 100, 500)  # smooth x-values
pdf = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-(x-mu)**2/(2*sigma**2)) 

# Scale PDF to histogram counts
bins=np.arange(0, 101, 1)
pdf_scaled = pdf * m 

plt.plot(x, pdf_scaled, 'r-', linewidth=2, label=f"Gaussian PDF)")

plt.xlabel('x')
plt.ylabel('Counts')
plt.title("Histogram of Gaussian Sample with PDF")
plt.xlim(0, 100)
plt.grid(True)
plt.legend()
plt.savefig("prob2c_cesarinegraham_histogram_with_pdf.png", dpi=300)
plt.show()

```

```python

m = 10000
mu = 55
sigma = 13

sample = np.random.normal(mu, sigma, size=m)

# bins from 0 to 100 with width 1
bins = np.arange(0, 101, 1)              # edges: 0,1,2,...,100
bin_width = bins[1] - bins[0]            # should be 1
bin_centers = bins[:-1] + bin_width/2    # centers: 0.5, 1.5, ..., 99.5


# Gaussian PDF evaluated at bin centers (continuous PDF)
pdf_at_centers = (1.0 / (sigma * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((bin_centers - mu)/sigma)**2)

# approximate counts in each bin by pdf(center) * N * bin_width
approx_counts = pdf_at_centers * m * bin_width

#plotting
plt.hist(sample, bins=bins, density=False, alpha=0.6, color='green', label='Sample counts')
plt.plot(bin_centers, approx_counts, color='red', linewidth=2, marker='o', markersize=4, label='Gaussian approx')

# formatting
plt.xlabel('x')
plt.ylabel('Counts')
plt.title(f'Histogram (N={m}, μ={mu}, σ={sigma}) with Gaussian overlay')
plt.xlim(0, 100)
plt.grid(True)
plt.legend()
plt.savefig("gaussian_histogram_bincenter_approx.png", dpi=300)
plt.show()
```
