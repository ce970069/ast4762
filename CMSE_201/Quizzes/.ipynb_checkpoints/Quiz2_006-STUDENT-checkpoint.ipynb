Now that you have a function to learn about the data, it's time to load in some data! The file, `endangered_species_cleaned.csv` has the total species assessed and endangered category data for mammals and amphibians. **Choose one species category (i.e. mammals or amphibians) to explore.**

**Using matplotlib, make a basic plot that visualizes the endangered category data on the y-axis and the years on the x-axis.** Make sure to include appropriate labels and a legend for full credit!

```python
# PUT YOUR CODE HERE
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

year, total, critically_endangered, endangered, vulnerable = np.loadtxt("endangered_species_cleaned.csv", usecols = [0,2,4,6,8], skiprows = 1, unpack=True, delimiter=',')

plt.figure(figsize=(8,6))
plt.plot(year,total, label = "Total Amphibian Species Assessed")
plt.plot(year,critically_endangered, label = "Total Amphibian Species Critically Endangered")
plt.plot(year,endangered, label = "Total Amphibian Species Endangered")
plt.plot(year,vulnerable, label = "Total Amphibian Species Vulnerable")
plt.title('Amphibian Species over the Years')
plt.xlabel('Years')
plt.ylabel('Number of Species')
plt.grid()
plt.legend()
```
