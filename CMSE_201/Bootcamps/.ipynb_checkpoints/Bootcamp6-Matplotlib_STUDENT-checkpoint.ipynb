# Bootcamp 6: Matplotlib
This notebook includes some basic examples of for using matplotlib!

```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
'''a basic plot example'''

x_values = np.linspace(0,4*math.pi,100) #this creates a numeric sequence from 0 to 4pi at 100 intervals
y_values = np.cos(x_values) #y is a cosine function of all the x values created above

plt.plot(x_values,y_values,label='my function') #plotting the values and laebling the line
plt.xlabel('x') #labeling the x axis
plt.ylabel('y') #labeling the y axis
plt.title('x vs. cos(x)') #giving the plot a title
plt.legend(loc='best') #creating a legend where it best fits
plt.grid() #adding the grid
plt.tight_layout() #adjusting the padding to best fit the data
```

```python
'''an example with plt.subplot()'''

x_values = np.linspace(0,4*math.pi,100) 
y_values = np.cos(x_values)
y2_values = x_values**2
y3_values = 2*x_values+1

plt.figure(1, figsize=(6,10)) #adjusting the figure size of each of the figures
plt.subplot(3,1,1) # write a comment to explain what this line is doing
plt.plot(x_values,y_values,label='cos(x)',color='springgreen',marker='o') #changing the dots and color of the plot
plt.xlabel('x') 
plt.ylabel('y') 
plt.legend(loc='lower center') #telling the plot where to put the legend

plt.subplot(3,1,2) #adding the new subplot
plt.plot(x_values,y2_values,label='x**2',color='lightblue',linestyle='--') # changing the dots and the color of the plot
plt.xlabel('x') 
plt.ylabel('y')
plt.legend(loc='upper left') # write a comment to explain what this line is doing

plt.subplot(3,1,3) # write a comment to explain what this line is doing
plt.plot(x_values,y3_values,label='2*x+1',color='blueviolet',linestyle='-.',linewidth=4) # changing the format of the line
plt.ylabel('y') 
plt.legend(loc='lower right',fontsize=12) #telling where to put the plot

plt.tight_layout()
```

```python
# put your code and pseudocode here

x_for_sin = np.arange(0,6*np.pi,0.2) # creating an array of values from 0 to 6pi at increments of 0.2
y_values = np.sin(x_for_sin) #y is a cosine function of all the x values created above

plt.plot(x_for_sin,y_values,label='sine(x)',color = 'darkgreen',linewidth = 3,marker = '*') #plotting the values and laebling the line
plt.xlabel('x') #labeling the x axis
plt.ylabel('sine(x)') #labeling the y axis
plt.title('x vs. sin(x)') #giving the plot a title
plt.legend(loc='best') #creating a legend where it best fits
plt.grid() #adding the grid
plt.tight_layout() #adjusting the padding to best fit the data
```

1. Write a function to calculate $y=ax^2+bx+c$
2. Using your function, calculate values for three different sets of a,b,c values. You may also use different x-values for each set if you choose.
3. Using plt.subplot(), plot the three results, each in different plot. Make sure to include a axis labels, a title, and a legend. Vary the colors, markers, line widths, etc. until you have a plots you like the look of!
4. Save your plot when you are done!

```python
# Put your pseudocode and code here

def first_plot(a,b,c):
    xvals = np.arange(0,5,0.1)
    for i in range(len(xvals)):
        y = a*xvals**2+b*xvals+c
        return y
```

```python
x = np.arange(0,5,0.1)
plt.plot(x, first_plot(1,2,3),label = 'f(x)',color = 'darkgreen',linewidth = 3,marker = '*')
plt.xlabel('x') #labeling the x axis
plt.ylabel('y') #labeling the y axis
plt.title('x vs. y') #giving the plot a title
plt.legend(loc='best') #creating a legend where it best fits
plt.grid() #adding the grid
plt.tight_layout() #adjusting the padding to best fit the data
```
