## 2. Modeling the Air Atop Everest

The higher you are above sea level, the thinner the air gets. For instance, in Denver (elevation: 1 mile, or 1600 meters), air density is about 12% lower than it is in East Lansing. For every eight breaths you take in Michigan, you’ll need nine breaths in Denver to get the equivalent amount of air.

This difference in air density with altitude can be lethal for mountain climbers, where the density of air can be a third or less the value at sea level. Meaning that for every one breath you take in Michigan, you’d need to take *three* breaths on top of Mount Everest to get the equivalent amount of air.

You are working for a company that specializes in selling air tanks to climbers. You’ve been tasked with writing a piece of code that can determine the density of air for a given altitude so that your company can calculate how many air tanks a climber will need. 

You don’t have much of a physics or chemistry background, so you’re not sure how to make a physical model for this situation. But you do have data from a previous expedition, where the climbers took measurements of the density at different heights. 

####  &#9989;&nbsp; **2.1** Use the data provided to test out different models. Decide on one model that you feel best fits the data. 

The three models you should test out are:

### Linear Model
$$y = Ax + B$$

### Quadratic Model
$$y = Ax^2 + Bx + C$$

### Exponential Model
$$y = A\exp(-Bx) + C$$


```python
#Data to use for your model
#Height is in kilometers, density is in kg/m^3

height = np.array([0.0,0.17,0.34,0.52,0.69,0.86,1.03,1.21,1.38,1.55,1.72,1.9,2.07,2.24,2.41,2.59,2.76,2.93,3.1,3.28,3.45,3.62,3.79,3.97,4.14,4.31,4.48,4.66,4.83,5.0])
density = np.array([1.2,1.23,1.19,1.2,1.17,1.12,1.07,1.05,1.06,1.04,1.03,1.01,0.95,0.97,0.96,0.93,0.93,0.88,0.87,0.9,0.83,0.84,0.84,0.8,0.79,0.8,0.78,0.77,0.73,0.76])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def linear_function(x, A, B):
    return A * x + B

def quadratic_function(x, A, B, C):
    return A*x**2 + B*x + C

def exponential_function(x, A, B, C):
    return A * np.exp(-B*x) + C

linear_fit, just_ignore_this = curve_fit(linear_function, height, density)
quadratic_fit, just_ignore_this = curve_fit(quadratic_function, height, density)
exponential_fit, just_ignore_this = curve_fit(exponential_function, height, density)

A_linear_expected = linear_fit[0]  
B_linear_expected = linear_fit[1]
y_expected_linear = linear_function(height, A_linear_expected, B_linear_expected)

A_quadratic_expected = quadratic_fit[0]  
B_quadratic_expected = quadratic_fit[1]
C_quadratic_expected = quadratic_fit[2]
y_expected_quadratic = quadratic_function(height, A_quadratic_expected, B_quadratic_expected, C_quadratic_expected)

A_exponential_expected = exponential_fit[0]  
B_exponential_expected = exponential_fit[1]
C_exponential_expected = exponential_fit[2]
y_expected_exponential = exponential_function(height, A_exponential_expected, B_exponential_expected, 
                                              C_exponential_expected)

plt.figure(figsize = (14,16))

plt.subplot(3,3,1)
plt.scatter(height,density, label = "data", color = "mediumseagreen")
plt.plot(height, y_expected_linear, color = "firebrick", label = "linear fit")
plt.title("linear model")
plt.legend()

plt.subplot(3,3,2)
plt.scatter(height,density, label = "data", color = "mediumseagreen")
plt.plot(height, y_expected_quadratic, color = "firebrick", label = "quadratic fit")
plt.title("quadratic model")
plt.legend()

plt.subplot(3,3,3)
plt.scatter(height,density, label = "data", color = "mediumseagreen")
plt.plot(height, y_expected_exponential, color = "firebrick", label = "exponential fit")
plt.title("exponential model")
plt.legend()
```

####  &#9989;&nbsp; **2.2** Code up the algorithm you defined in part 1 and *quantitatively* compare how well each model fits the data.

```python
#Write your code for goodness of fit here

def best_best_fit(line_y, act_y):
    holder = 0
    for i in range(len(act_y)):
        holder += line_y[i] - act_y[i]
    return holder/len(act_y)


print("Linear test:", best_best_fit(y_expected_linear, density))
print("Quadratic test:", best_best_fit(y_expected_quadratic, density))
print("Exponential test:", best_best_fit(y_expected_exponential, density))
```

####  &#9989;&nbsp; **2.3** Which model best fits the data? Is it substantially better than the other models?

The linear model fits the data the best

```python
answer = linear_function(8.85, linear_fit[0], linear_fit[1])
answer2 = quadratic_function(8.85, quadratic_fit[0], quadratic_fit[1], quadratic_fit[2])
answer3 = exponential_function(8.85, exponential_fit[0], exponential_fit[1], exponential_fit[2])

print("Linear prediction:", answer)
print("Quadratic prediction:", answer2)
print("Exponential prediction:", answer3)
```
