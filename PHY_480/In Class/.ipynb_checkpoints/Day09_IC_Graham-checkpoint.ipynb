## Part 1: Quick Recap: The Trapezoidal Rule and Simpson's Rule

Suppose we have a function $f(x)$ and we want to calculate its integral with respect to $x$ from $x=a$ to $x=b$, which we denote $I(a,b)$

$$ I(a,b) = \int_a^b dx\, f(x), $$

We will use the trapezoidal rule and Simpson's rule to calculate the area under the curve of $f(x)$ from $a$ to $b$ using different approximations.

### 1.1: Trapezoidal rule

In the trapezoidal rule, we use a straight line to approximate the area under the function.
We divide the interval from  $a$ to $b$ into $N$ slices, so that each slice has width $h=\frac{b-a}{N}$.
The $k$th slice's part of the interval goes from $a+(k-1)h$ to $a+kh$;
thus, its area is

$$ A_k = \frac{1}{2}h \left[f(a+(k-1)h)+f(a+kh)\right]. $$

The approximation of the area under the whole curve is the sum of the areas of the trapezoids for all $N$ slices:

$$ I(a,b) \approx \sum_{k=1}^N A_k =  h\left[\frac{1}{2}f(a)+\frac{1}{2}f(b)+ \sum_{k=1}^{N-1} f(a+kh)\right]. $$

In the cell below, write a function called `trapezoidal_rule(f, N, a, b)` that takes a function `f(x)`, $N$ slices over an interval from $a$ to $b$, and returns the integral approximated using the trapezoidal rule.


```python
def trapezoidal_rule(f, N, a, b):
    h = (b - a) / N # calculating the width of each interval
    integral = 0.5 * (f(a) + f(b))  # trapezoidal rule formula for endpoints
    for k in range(1, N): # summing the function values at each interior point
        integral += f(a + k * h)  # adding the value of each slice
    integral *= h  # multiplying the sum by the width of each subinterval
    return integral  # returning the final result
```

###  1.3: Simpson's rule

Simpson's rule uses a quadratic curve to approximate the area under the function; to do so, one needs 3 points (instead of 2 as in the trapezoidal rule case).

Again, we divide the interval from  $a$ to $b$ into $N$ slices, so that each slice has width $h=\frac{b-a}{N}$.
The $k$th slice's part of the interval goes from $a+(k-1)h$ to $a+kh$;
thus, its area is

$$ A_k= \frac{1}{3}h \left[f(a+k h)+ 4 f(a+(k+1)h) + f(a+(k+2)h) \right]. $$

Note that the total number of slices must be _even_ for this to work.
Collecting the terms together, we now have

$$\begin{align}
I(a,b) &\approx \sum_1^N A_k \\
       &=  \frac{1}{3}h \left[f(a) + 4 f(a+h) + 2 f(a+2h) + 4 f(a+3h) + \ldots  +f(b) \right] \\
       &=  \frac{1}{3}h \left[f(a)+f(b) + 4 \sum_{\text{odd } k=1}^{N-1} f(a+k h) + 2 \sum_{\text{even }k= 2}^{N-2} f(a+ kh)  \right]
\end{align}$$

```python
def simpson_rule(f, N, a, b):

    if N % 2 != 0:
        raise ValueError("Number of slices (N) must be even for Simpson's rule.")
    
    h = (b - a) / N  # returning the final result
    integral = f(a) + f(b)  # initializing integral with endpoints
    
    # Adding the value of each slice to the integral
    for k in range(1, N): # for ODD numbers
        x = a + k * h
        if k % 2 == 0: # for EVEN numbers
            integral += 2 * f(x)  # Even-indexed terms multiplied by 2
        else:
            integral += 4 * f(x)  # Odd-indexed terms multiplied by 4
    
    integral *= h / 3 # multiplying the sum by the width and divide by 3
    return integral # returning the final result
```

## Part 2: Diffraction Limit of a Telescope

Our ability to resolve detail in astronomical observations is limited by the diffraction of light in our telescopes.
Light from stars can be treated effectively as coming from a point source at infinity.
When such light with wavelength $\lambda$ passes through the circular aperture of a telescope (which we'll assume to have unit radius) and is focused by the telescope on the focal plane, it produces not a single dot, but a  circular diffraction pattern consisting of central spot surrounded by a series of concentric rings.
The intensity of the light in this diffraction pattern is given by

$$ I(r) = \left( {J_1(kr)\over kr} \right)^2,$$

where $r$ is the distance in the focal plane from the center of the diffraction pattern, $k=2\pi/\lambda$, and $J_1(x)$ is a Bessel function.

The Bessel functions $J_m(x)$ are given by

$$J_m(x) = {1\over\pi} \int_0^\pi d\theta\, \cos(m\theta - x\sin\theta),$$

where $m$ is a nonnegative integer and $x \ge 0$.

### 2.1
Write a Python function `bessel_J(m, x)` that calculates the value of $J_m(x)$ using Simpson's rule with $N=1000$ points.
Use your function in a program to make a plot, on a single graph, of the Bessel functions $J_0$, $J_1$, and $J_2$ as functions of $x$ from $x=0$ to $x=20$.

```python
def bessel_J(m, x):
    n = 1000
    theta = np.linspace(0, np.pi, n)
    integral = simpson_rule(lambda t: np.cos(m*t - x*np.sin(t)), n, 0, np.pi)
    return integral/np.pi

m = [0, 1, 2]
x = np.linspace(0, 20, 1000)
J0 = []
J1 = []
J2 = []

for i in range(len(x)):
    J0.append(bessel_J(m[0], x[i]))
    J1.append(bessel_J(m[1], x[i]))
    J2.append(bessel_J(m[2], x[i]))
    
plt.plot(x, J0)
plt.plot(x, J1)
plt.plot(x, J2)
plt.grid()
```
