The general multistate fit form is

## Part 3: Calculate effective masses

Some of you may notice that since $M_1$ is much larger than $M_0$ and $A_1$ is also smaller than $A_0$, at larger $t$, the $C_\text{pion}(t)$ will be dominated by the first term: $A_0\exp(-M_0 t)$.
This allows us to find easy way to "eyeball" what the ground-state mass is using the *effective mass*:

$$M_\text{eff}(t) = \ln \frac{C(t)}{C(t+1)}$$.

For those who has not seen "ln" before, it is the "natural log", the inverse function of "exp".
If you already loaded the math module, you can use `math.log()` for this.
