## Part 2:   Wave Equation in 2+1 Dimensions

### 2.1 Start with 1+1 dimension
The $1+1$-dimensional wave equation is

$$ \frac{\partial^2 u}{\partial x^2}=\frac{\partial^2 u}{\partial t^2}, $$

with $u=u(x,t)$, and we operate with dimensionless variables (that is, we lazily left out some factors of the wave speed $c$). Possible boundary and initial conditions with $L=1$ are

$$\begin{array}{cc}
    u(x,0) = g(x)& x\in (0,1) \\
    u(0,t)=u(1,t)=0 & t > 0 \\
    \partial u/\partial t|_{t=0}=0 & x\in (0,1)\\
\end{array}.$$

#### Discretization
We discretize again in time and position:

$$\begin{align}
\frac{\partial^2 u}{\partial x^2} &\approx \frac{u(x+\Delta x,t)-2u(x,t)+u(x-\Delta x,t)}{\Delta x^2} \\
\frac{\partial^2 u}{\partial t^2} &\approx \frac{u(x,t+\Delta t)-2u(x,t)+u(x,t-\Delta t)}{\Delta t^2},
\end{align}$$

which we rewrite as

$$\begin{align}
\frac{\partial^2 u}{\partial x^2} &\approx \frac{u_{i+1,j}-2u_{i,j}+u_{i-1,j}}{\Delta x^2} \\
\frac{\partial^2 u}{\partial t^2} &\approx \frac{u_{i,j+1}-2u_{i,j}+u_{i,j-1}}{\Delta t^2},
\end{align}$$

resulting in

$$ u_{i,j+1} = 2u_{i,j} - u_{i,j-1} + \frac{\Delta t^2}{\Delta x^2}\left(u_{i+1,j}-2u_{i,j}+u_{i-1,j}\right). $$

If we assume that all values at times $t=j$ and $t=j-1$ are known, the only unknown variable is $u_{i,j+1}$ and the last equation yields thus an explicit scheme for updating this quantity.
We have thus an explicit finite difference scheme for computing the wave $u$.

#### Note on the initial timestep

The only additional complication in our case is the initial condition given by the first derivative in time,
namely $\partial u/\partial t|_{t=0}=0$.
The discretized version of this first derivative is given by

$$ \frac{\partial u}{\partial t} \approx \frac{u(x_i,t_j+\Delta t)-u(x_i,t_j-\Delta t)}{2\Delta t}, $$

and at $t=0$ it reduces to

$$ \frac{\partial u}{\partial t} \approx \frac{u_{i,+1}-u_{i,-1}}{2\Delta t}=0, $$

implying that $u_{i,+1} = u_{i,-1}$.

If we insert this condition into the previous formula, we get a special formula for the first timestep

$$ u_{i,1} = u_{i,0}+\frac{\Delta t^2}{2\Delta x^2}\left(u_{i+1,0}-2u_{i,0}+u_{i-1,0}\right). $$

We seemingly need two different equations, one for the first timestep and one for all other timesteps.
However, we can avoid this complication by setting

$$ u_{i,-1}=u_{i,0}+\frac{\Delta t^2}{2\Delta x^2}\left(u_{i+1,0}-2u_{i,0}+u_{i-1,0}\right), $$

in our setup of the initial conditions.
Then, we can use the ordinary evolution step for all timesteps.

### 2.2 Wave equation in 2+1 dimensions

The situation is rather similar for the $1+1$-dimensional case, except that we now need to discretize the spatial $y$-coordinate as well.
Our equations will now depend on three variables whose discretized versions are now

$$
\begin{array}{cc}
    t_l=l\Delta t& l \ge 0 \\
    x_i=i\Delta x& 0 \le i \le n_x \\
    y_j=j\Delta y& 0 \le j \le n_y
\end{array} ,
$$

and we will let $\Delta x=\Delta y = h$ and $n_x=n_y$ for the sake of simplicity.
The initial and boundary conditions are

$$\begin{array}{cc}
  u(x,y,0) = g(x,y)& x,y\in (0,1) \\
  u(0,0,t) = u(1,1,t)=0 & t > 0 \\
  \partial u/\partial t|_{t=0} = 0 & x,y\in (0,1) \\
\end{array}.$$

We have now the following discretized partial derivatives

$$\begin{align}
\frac{\partial^2 u}{\partial x^2} &\approx \frac{u_{i+1,j}^l-2u_{i,j}^l+u_{i-1,j}^l}{h^2} \\
\frac{\partial^2 u}{\partial y^2} &\approx \frac{u_{i,j+1}^l-2u_{i,j}^l+u_{i,j-1}^l}{h^2} \\
\frac{\partial^2 u}{\partial t^2} &\approx \frac{u_{i,j}^{l+1}-2u_{i,j}^{l}+u_{i,j}^{l-1}}{\Delta t^2}, \\
\end{align}$$

which we merge into the discretized $2+1$-dimensional wave equation
as

$$
u_{i,j}^{l+1} = 2u_{i,j}^{l}-u_{i,j}^{l-1} +
  \frac{\Delta t^2}{h^2}\left(u_{i+1,j}^l-4u_{i,j}^l + u_{i-1,j}^l + u_{i,j+1}^l + u_{i,j-1}^l\right),
$$

where we have an explicit scheme with $u_{i,j}^{l+1}$ as the only unknown quantity.

The partial derivative is treated in much the same way as for the one-dimensional case, except that we now have an additional index due to the extra spatial dimension, so we need to compute $u_{i,j}^{-1}$ through

$$
u_{i,j}^{-1}=u_{i,j}^0+\frac{\Delta t}{2h^2}\left(u_{i+1,j}^0-4u_{i,j}^0+u_{i-1,j}^0+u_{i,j+1}^0+u_{i,j-1}^0\right),
$$

to set up the initial conditions.
