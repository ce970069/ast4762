# Part 1: Monte-Carlo Integration

Whereas techniques such as the trapezoidal rule or Gaussian quadrature use a deterministic approach, Monte-Carlo integration simplifies the integration process while introducing random noise, using large numbers of random samples to approximate the actual solution of an integral.
Consider an integral being taken for a function $f(\mathbf{x})$ over a domain $\Omega$ with volume $V$:

$$ V = \int_\Omega d\mathbf{x} $$

Then, we can take $N$ samples selected from a uniform distribution over that domain and evaluate the function at those points:

$$ \int_\Omega d\mathbf{x}\,f(\mathbf{x}) \approx \frac{V}{N} \sum_{i=1}^N f(\mathbf{x}_i) $$

# Part 2: Probability Distribution Function

The _probability distribution function_ (PDF) is a function $p(x)$ on the domain which, in the discrete case, gives us the probability or relative frequency with which these values of $X$ occur

$$ p(x) = \text{Prob}(X=x). $$

In the continuous case, the PDF does not directly depict the actual probability.
Instead, we define the probability for the stochastic variable to assume any value on an infinitesimal interval around $x$ to be $p(x)dx$.
The continuous function $p(x)$ then gives us the density of the probability rather than the probability itself.
The probability for a stochastic variable to assume any value on a non-infinitesimal interval $[a,b]$ is then just the integral

$$ \text{Prob}(a\leq X\leq b) = \int_a^b dx\,p(x). $$

Qualitatively speaking, a stochastic variable represents the values of numbers chosen as if by chance from some specified PDF so that the selection of a large set of these numbers reproduces this PDF.

The produced figure  shows the development of this system as function of time steps.
We note that for $N=1000$ after roughly $2000$ time steps, the system has reached the equilibrium state.
There are, however, fluctuations around equilibrium.

The following table collects properties of probability distribution functions.
In our notation we reserve the label $p(x)$ for the probability of a certain event, while $P(x)$ is the cumulative probability.

| | Discrete PDF | Continuous PDF |
| ---: | :---: | :---: |
| Domain  | $\left\{x_1, x_2, x_3, \dots, x_N\right\}$  | $[a,b]$  |
| Probability  | $p(x_i)$  | $p(x)dx$  |
| Cumulative  | $P_i=\sum_{l=1}^ip(x_l)$  | $P(x)=\int_a^xp(t)dt$  |
| Positivity  | $ 0\le p(x_i)\le 1$  | $ p(x) \ge 0$  |
| Positivity  | $ 0\le P_i\le 1$  | $ 0\le P(x)\le 1$  |
| Monotonic  | $P_i\ge P_j$ if $x_i\ge x_j$  | $P(x_i)\ge P(x_j)$ if $x_i\ge x_j$  |
| Normalization  | $P_N=1$  | $P(b)=1$  |
| | <span style="color:transparent">--------------------------------</span> | <span style="color:transparent">-----------------------------</span> |

With a PDF we can compute expectation values of selected quantities such as the moments

$$ \langle x^k\rangle=\frac{1}{N}\sum_{i=1}^{N}x_i^kp(x_i), $$

if we have a discrete PDF or

$$ \langle x^k\rangle=\int_a^b dx\,x^k p(x), $$

in the case of a continuous PDF.
Recall that the mean value $\mu$ and the variance $\sigma^2$ can be expressed in terms of such moments.

## Three famous PDFs

Although there are innumerable PDFs even if you limit yourself to those with names, there are a few that tend to turn up much more often than the others.

### Uniform distribution

$$ p(x)=\frac{1}{b-a}\Theta(x-a)\Theta(b-x) $$

Here, $\Theta$ is the step function.
The uniform distribution is nonzero on the interval $[a,b]$ and zero elsewhere.
Random-number generators usually produce random numbers of uniform distribution.
Dice and wheels produce random numbers with a discrete uniform distribution.

### Normal distribution

$$ p(x)=\frac{1}{b\sqrt{2\pi}}\exp{\left(-\frac{(x-a)^2}{2b^2}\right)} $$

Also known as the _Gaussian distribution_, this PDF is nonzero for all real numbers.

When computing the moments of this PDF, it's useful to know the relation $\int_{-\infty}^{\infty}dx\,\exp\left(-x^2\right) = \sqrt{\pi}$.

The mean value

$$\begin{align}
\mu &= \int_0^{\infty}dx\,xp(x) \\
    &= \frac{1}{b\sqrt{2\pi}}\int_{-\infty}^{\infty}dx\,x \exp\left(-\frac{(x-a)^2}{2b^2}\right) \\
    &= \frac{1}{b\sqrt{2\pi}}\int_{-\infty}^{\infty}dy\,b\sqrt{2}\left(a+b\sqrt{2}y\right)\exp\left(-y^2\right) \\
    &=a.
\end{align}$$

Similarly, the variance is

$$\begin{align}
\sigma^2 &= \frac{1}{b\sqrt{2\pi}}\int_{-\infty}^{\infty}dx\,(x-\mu)^2 \exp\left(-\frac{(x-a)^2}{2b^2}\right) \\
         &= \frac{1}{b\sqrt{2\pi}}\int_{-\infty}^{\infty}dy\,b\sqrt{2}\left(b\sqrt{2}y\right)^2\exp\left(-y^2\right) \\
         &= \frac{2b^2}{\sqrt{\pi}}\int_{-\infty}^{\infty}dy\,y^2\exp\left(-y^2\right) \\
         &= b^2
\end{align}$$

The "standard" normal distribution is defined by $\mu=a=0$ and variance $\sigma^2=1$, leading to

$$ p(x)=\frac{1}{\sqrt{2\pi}}\exp\left(-\frac{x^2}{2}\right). $$

### Exponential distribution

$$ p(x)=\alpha e^{-\alpha x} $$

This PDF is nonzero for positive real numbers and has mean

$$\begin{align}
\mu &= \int_0^{\infty}dx\,xp(x) \\
    &= \int_0^{\infty}dx\,x\alpha e^{-\alpha x} \\
    &=\frac{1}{\alpha}
\end{align}$$

with variance

$$\begin{align}
\sigma^2 &= \int_0^{\infty}dx\,x^2p(x)-\mu^2 \\
         &= \frac{1}{\alpha^2}.
\end{align}$$

## Cumulative probability
We will also often see the _cumulative probability distribution function_ (CDF), denoted $P(x)$, which is just the probability for a stochastic variable $X$ to assume any value less than $x$

$$
P(x)=\text{Prob(}X\leq x\mathrm{)} =
\int_{-\infty}^x dx^\prime\, p\left(x^{\prime}\right).
$$

The relation between a CDF and its corresponding PDF is then

$$ p(x) = \frac{d}{dx}P(x). $$
