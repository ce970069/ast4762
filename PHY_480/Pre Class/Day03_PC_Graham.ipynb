### 1.1 Plot two-dimensional data
Make a simple plot with the data given below, using `x` for the horizontal axis and `y` for the vertical axis.


### 1.2 Plot a sine wave
Generate a list of samples along the horizontal axis ranging $x \in [0, 10]$ with total 100 points.
Generate $y=\sin(x)$ and plot on the vertical axis.
Remember to label the axes appropriately.

```python
# put your code below

# Generate 100 points between 0 and 10
x_values = np.linspace(0, 10, 100)

# Calculate y = sin(x) for each x
y_values = np.sin(x_values)

# Plotting
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y = sin(x)')
plt.title('Sine Function Plot')
plt.grid(True)
plt.show()
```


### 1.3 Add a cosine wave
Using the same $x$ samples from 1.2, add $y_2=\cos(x)$ to the plot in a different color.
(That is, you should have one plot with both $\sin(x)$ and $\cos(x)$ in it.)
Remember to label the axes and add a legend describing how the colors correspond to the data.

```python
# put your code below

# Generate 100 points between 0 and 10
x_values = np.linspace(0, 10, 100)

# Calculate y = sin(x) and y = cos(x) for each x
y_sin = np.sin(x_values)
y_cos = np.cos(x_values)

# Plotting
plt.plot(x_values, y_sin, label='sin(x)', color='blue')
plt.plot(x_values, y_cos, label='cos(x)', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Functions Plot')
plt.legend()
plt.grid(True)
plt.show()
```

### 1.4 Log plot
Using the same $x$ samples from 1.2, calculate $y_3=2\exp(-0.5x)$.
Create two plots: for both use a linear horizontal scale for $x$; for one use a linear vertical scale for $y$ and for the second use a logarithmic vertical scale.  
Remember to label the axes and add a legend.

```python
# put your code below

# Generate 100 points between 0 and 10
x_values = np.linspace(0, 10, 100)

# Calculate y3 = 2 * exp(-0.5x) for each x
y3_values = 2 * np.exp(-0.5 * x_values)

# Plot with linear vertical scale for y
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x_values, y3_values, label='2 * exp(-0.5x)', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Vertical Scale')
plt.legend()
plt.grid(True)

# Plot with logarithmic vertical scale for y
plt.subplot(1, 2, 2)
plt.plot(x_values, y3_values, label='2 * exp(-0.5x)', color='green')
plt.xlabel('x')
plt.ylabel('y (log scale)')
plt.yscale('log')  # Set logarithmic scale for y-axis
plt.title('Logarithmic Vertical Scale')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Part 2: Deltoid curve


Although the `plot` function is designed primarily for plotting functions on the vertical axis sampled along the horizontal axis, it can be adapted for other kinds of plotting as well.

Make a plot of the so-called _deltoid_ curve, which is defined parametrically by the equations

$x = 2 \cos \theta + \cos 2\theta $

$y = 2 \sin \theta - \sin 2\theta $

where $0\le\theta<2\pi$.  

### 2.1 Parametric plot
Sample $\theta$ between zero and $2\pi$ and calculate $x$ and $y$ for each sample using the equations above.
Then, plot $y$ as a function of $x$.

```python
# put your code below

# Sample theta between 0 and 2*pi
theta_values = np.linspace(0, 2 * np.pi, 1000)

# Parametric equations for the deltoid curve
x_values = 2 * np.cos(theta_values) + np.cos(2 * theta_values)
y_values = 2 * np.sin(theta_values) - np.sin(2 * theta_values)

# Plotting y as a function of x
plt.plot(x_values, y_values, label='Deltoid Curve', color='purple')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Deltoid Curve: y as a function of x')
plt.legend()
plt.grid(True)
plt.show()
```


### 2.2 Polar plot, if time allows
Taking this approach a step further, one can make a polar plot $r=f(\theta)$ for some function $f$ by calculating $r$ for a range of values of $\theta$
and then converting $r$ and $\theta$ to Cartesian coordinates using the standard equations

$x = r\cos\theta$,

$y = r\sin\theta$.  

Use this method to make a plot of the Galilean spiral $r = \theta^2$ for $0\le\theta\le 10\pi$.

```python
# put your code below

# Sample theta between 0 and 10*pi
theta_values = np.linspace(0, 10 * np.pi, 100)

# Polar equation for the Galilean spiral
r_values = theta_values**2

# Convert polar coordinates to Cartesian coordinates
x_values = r_values * np.cos(theta_values)
y_values = r_values * np.sin(theta_values)

# Plotting the Galilean spiral in polar coordinates
plt.polar(theta_values, r_values, label='Galilean Spiral', color='m')
plt.title('Galilean Spiral: Polar Plot')
plt.legend()
plt.show()
```

### 2.3  Fey's function, if time allows
Using the same method, make a polar plot of Fey's function:

$r = e^{\cos\theta} - 2 \cos 4\theta + \sin^5 \frac{\theta}{12}$

over the range $0\le\theta\le 24\pi$.

```python
# put your code below

# Sample theta between 0 and 24*pi
theta_values = np.linspace(0, 24 * np.pi, 1500)

# Fey's function in polar coordinates
r_values = np.exp(np.cos(theta_values)) - 2 * np.cos(4 * theta_values) + np.sin(theta_values/12)**5

# Convert polar coordinates to Cartesian coordinates
x_values = r_values * np.cos(theta_values)
y_values = r_values * np.sin(theta_values)

# Plotting Fey's function in polar coordinates
plt.plot(y_values, x_values, label="Fey's Function", color='m')
plt.title("Fey's Function: Polar Plot")
plt.grid()
plt.legend()
plt.show()
```
