## Part 1: Introduction

Some processes in physics are random, like radioactive decay.
Quantum mechanics tells use the probability of decay per unit time, but the exact moment of decay is random and cannot be calculated.
Other physical processes are not truly random, but we can treat them as random for all purposes.
For example, Brownian motion; if we knew the exact position and velocity of every molecule, we could predict Brownian motion.
But in practice we cannot keep track of every single molecule, so many-body systems are often better treated as random processes.
In this class, we will look at examples of computational methods for studying random physical processes, both truly random and apparently random.
To start, we should first learn about random-number generators.

### 1.1 Linear congruential random-number generator

Consider the following equation

$$ x^\prime = (a x +c) \text{ mod } m $$

where $a$, $c$, and $m$ are constant integers, and $x$ is an integer variable.

For suitably selected constants, this generates a string of apparently random integers, simply by iterating the same equation over and over again.
Here is a program to do so:

```python
from pylab import plot, show

N = 1000
a = 1664525
c = 1013904223
m = 4294967296
x = 1
results = []

for i in range(N):
    x = (a*x + c)%m
    results.append(x)
plot(results,"o")
show()
```

It should be obvious that this program is not truly random, since it is deterministic.
If we run the same program with the same constants, we get the same "random" numbers.
However, for a physical system with a small number of degrees of freedom, these numbers are sufficiently random for the calculation.

### 1.4 Random number seeds
By definition of the random-number generator, you will always get a different number every time you sample from the random-number function.
However, sometimes we want to be able to reproduce the results from an earlier program to double check the answer.
For this purpose, you can use a selected random-number _seed_ to make sure that the output is repeatable.
Run the following code _multiple times_ to look for differences in the output, and write down your observation.

```python
# without seed
from random import randrange, seed
for i in range(4):
    print(randrange(10))

print("run the same code for a second time")
for i in range(4):
    print(randrange(10))
```

```python
# with seed
from random import randrange, seed
seed(42)
for i in range(4):
    print(randrange(10))

print("run the same code for a second time")
seed(42)
for i in range(4):
    print(randrange(10))
```

<font color='blue'>
Without a seed, each time you run the code you get a different set of random numbers. With a seed, each time you run the code you get the same set of random numbers.

# Part 2: Random Walk  

Random walks are sometimes referred to as [Brownian motion](https://en.wikipedia.org/wiki/Brownian_motion), which referred originally to the motions of a single grain of pollen suspended in a liquid solution.
It is named after Robert Brown, who studied pollen under a microscope in 1829 in an attempt to understand the reproductive process in plants.
The Roman philosopher Lucreitus predicted this phenomenon much earlier in a poem from over 200 years ago:

>"Observe what happens when sunbeams are admitted into a building and shed light on its shadowy places. You will see a multitude of tiny particles mingling in a multitude of ways... their dancing is an actual indication of underlying movements of matter that are hidden from our sight... It originates with the atoms which move of themselves. Then those small compound bodies that are least removed from the impetus of the atoms are set in motion by the impact of their invisible blows and in turn cannon against slightly larger bodies. So the movement mounts up from the atoms and gradually emerges to the level of our senses, so that those bodies are in motion that we see in sunbeams, moved by blows that remain invisible."

(Actually, dust motes move due to tiny air currents, but he had the spirit of the argument correct.)
Nonetheless, Brownian motion remained surprisingly challenging to understand until Einstein in 1905, who rigorously tied the motion of the large particles (pollen) to the motions of many atoms (the molecules of the water the pollen was in).
What Brown saw under his microscope was a jiggling pollen grain, and Einstein showed that the jitter was due to many rapid collisions with water molecules.
![Brown](https://upload.wikimedia.org/wikipedia/commons/c/c2/Brownian_motion_large.gif)

### 2.1

Let's start with the simplest case: a 1D random walk; that is, a random walk on a line.

Imagine that you draw a line on the floor, with a mark every foot.
You start at the middle of the line (the point you decide is the origin, at $x=0$).
You then flip a fair coin $N_\text{steps}$ times ("fair" means that it has equal chances of coming up heads or tails).
Every time the coin comes up heads, you take one step to the right (in the $+$ direction).
Every time it comes up tails, you take a step to the left (in the $-$ direction).  

**The questions we're going to answer:**

* After $N_\text{steps}$ coin flips and steps, how far are you from the origin?
* If you repeat this experiment $N_\text{trials}$ times, what will the distribution of distances from the origin be, and what is the mean distance that you go from the origin?  (Note: "distance" means the absolute value of displacement from the origin!)

Write a function, called `step1d`, that generates a random integer, `1` for heads and `0` for tails, and then **returns** instructions to take a step either left or right, which means that it should return `-1` or `1`.  
Call the function a few times and print out the values it returns to make sure it works; we'll use this as the basis of our random walk.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import random
```

```python
### Fill in the missing code below
def step1d():
    coin_flip = random.randint(0,1)             # use randint to give number between 0 and 1
    if coin_flip == 0:         # if coin_flip is 0 (for tail), and turn left (-1)
        step = -1
    elif coin_flip ==1:          # if coin_flip is 1 (for head), and turn right (+1)
        step = 1
    return step
```

Try to work on this function on your own; if you spend more than 15 minutes, see the end of the notebook to get the correct answer. You need this function to complete the following assignment.

```python
## Now test your function

steps = []
for i in range(10):
   steps.append(step1d())
print(steps)
```

### 2.2
Write a code in the space provided below to do a _single_ trial of the random walk with $N_\text{steps} = 1000$ using the function you have created.
Start at $x=0$ and take $N_\text{steps}$ steps to the right (when `step1d()` returns `1`) or left (when `step1d()` returns `-1`).

```python
# fill in the missing code

nsteps = 1000               # how many total steps do I want to take?
x = 0                       # Define starting position
for i in range(nsteps):     # loop over steps
    x = x + step1d()        # Take a step
print(x)
```

### 2.3
Copy the code from above and modify it to keep track of your distance from the origin as you take the step, and then plot displacement as a function of step.

_Hint: Start with empty list, and use `append()` to keep track of the `x` position._

```python
# your code here
nsteps = 1000               # how many total steps do I want to take?
x = 0                       # Define starting position
displacement = [x]          # List to store displacement at each step

for i in range(nsteps):     # loop over steps
    x = x + step1d()        # Take a step
    displacement.append(x)  # Append the new position to the list

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(range(nsteps + 1), displacement)
plt.xlabel('Step')
plt.ylabel('Displacement from Origin')
plt.title('1D Random Walk - Displacement vs. Step')
plt.grid(True)
plt.show()
```

### 2.4
Now, we want to see how several of these trajectories look at the same time.  
Copy and paste your code from above and add an outer loop where you make five separate trials,
and plot each of the trajectories on the same plot.  

You should get something that ends up looking like [this](https://drive.google.com/open?id=1_MUKk1RH5lmCs1BTpzaa7h0dbTstKR0n).

```python
# your code here
ntrials = 5             # Number of trials
nsteps = 1000           # Number of steps per trial
plt.figure(figsize=(10, 6))

for _ in range(ntrials):
    x = 0
    displacement = [x]
    for i in range(nsteps):
        x = x + step1d()
        displacement.append(x)
    plt.plot(range(nsteps + 1), displacement)

plt.xlabel('Step')
plt.ylabel('Displacement from Origin')
plt.title('1D Random Walk - Multiple Trials')
plt.grid(True)
plt.show()
```
