### 1.4 Running average of the data
Modify your program further to calculate and plot the running average of the data, defined by

$$Y_k = {1\over2r} \sum_{m=-r}^r y_{k+m}.$$

Use $r=5$ (and the $y_k$ are the sunspot numbers).

### 1.5 Replotting
Plot both the original data and the running average on the same graph, again over the range covered by the first 1000 data points.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.1

data = np.loadtxt("sunspots.txt", delimiter='	')
time = data[:, 0].astype(int) # changing this to integer since that is what the data type is
sunspot = data[:, 1]

print("Time:", time)
print("Sunspot:", sunspot)
```

#### 1.1

I started by importing NumPy. I then used np.loadtxt to read the text file into a NumPy array. After, I specified the file path, and delimiter in order to help the library sort the data. I then extracted the columns by separating the 'time' and 'sunspot' columns from the NumPy array.

```python
#1.2

plt.figure(figsize=(10, 6))
plt.plot(time, sunspot, marker='o', linestyle='-', color='m')
plt.title('Sunspot Number Over Time')
plt.xlabel('Time')
plt.ylabel('Sunspot Number')
plt.grid(True)
plt.show()
```

```python
#1.3

time_1000 = time[:1000]
sunspot_1000 = sunspot[:1000]

plt.figure(figsize=(10, 6))
plt.plot(time_1000, sunspot_1000, marker='o', linestyle='-', color='m')
plt.title('Sunspot Number Over Time (First 1000 Data Points)')
plt.xlabel('Time')
plt.ylabel('Sunspot Number')
plt.grid(True)
plt.show()
```

```python
#1.4

# defining the running average function
def running_average(data, r):
    n = len(data)
    y_running_avg = np.zeros_like(data, dtype=float)
    for k in range(n):
        start_idx = max(0, k - r) # starting the index at the max
        end_idx = min(n, k + r + 1) # ending the index at the min
        y_running_avg[k] = np.sum(data[start_idx:end_idx]) / (2 * r)
    return y_running_avg

# calculating running average with r=5
r_value = 5
sunspot_running_avg = running_average(sunspot, r_value)

# plotting
plt.figure(figsize=(10, 6))
plt.plot(time, sunspot, label='Sunspot Data', marker='o', linestyle='-', color='m')
plt.plot(time, sunspot_running_avg, label=f'Running Average (r={r_value})', linestyle='--', color='k')
plt.title('Sunspot Number and Running Average Over Time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

## Part 2: QCD data

One of the frontiers of research in particle and nuclear physics is to understand how fundamental particles, the gluons and quarks, determine the properties of ordinary matter. The lightest particle that can be made by these fundamental particles is called the "pion", and it is made up of one quark and one antiquark. However, in nature pions decay quickly; therefore, even though pions have lower mass than nucleons, the matter that makes up everyday objects is not made of pions. However, the properties of pions determine many important aspects of particle and nuclear physics.

In the Standard Model of particle physics, quantum chromodynamics (QCD), describes how gluons strongly interact with themselves and with quarks, forming (among other things) the pions. Using a sufficiently large supercomputer, we can use lattice QCD to simulate a tiny volume of spacetime containing a pion and study its properties. One output of such a simulation are *correlators* $C(t)$.
In this part of the exam, we are going to study the mass of the pion.

First: Download the correlator file, `pion.D-579.DG5p5_1.DG5p5_1.SS`[https://drive.google.com/file/d/1MBPJ9QgyBZu0JwmFpebvvev7hoCq3I7b/view?usp=sharing]

`pion.D-579.DG5p5_1.DG5p5_1.SS` has 3 columns of data, separated by spaces.

The first column of the data gives the time in very small time units. The second and third columns of data give the numerical value of the real and imaginary parts of pion correlator at the time indicated.
