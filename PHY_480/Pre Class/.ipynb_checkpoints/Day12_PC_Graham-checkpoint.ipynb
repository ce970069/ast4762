## Part 0: Introduction
The most common use of computers in physics is to solve differential equations.
Here, we will look at techniques for solving ordinary differential equations (ODEs).
We will sample a few of the most commonly taught methods in a typical computational-physics course, starting with single-variable ODEs: Euler's method, Runge-Kutta method, then moving to multi-variable ODEs and higher-order ODEs.

The general form of a first-order one-variable ODE is

$$ \frac{dx}{dt} = f(x,t),$$

where the $f(x,t)$ is the given physics relation.

We can write the value of $x$ a short interval $h$ later using a Taylor expansion; thus,

$$\begin{align}
x(t+h) &= x(t) + h  \frac{dx}{dt} + \frac{1}{2}h^2 \frac{d^2}{dt^2}+\ldots \\
       &=x(t) + h f(x,t) + O(h^2)
\end{align}$$

In Euler's method, we will ignore the  $O(h^2)$ (assuming $h$ is small enough) and

$$ x(t+h) \approx x(t) + h f(x,t)$$

### 1.2 Test on an example
Let's use Euler's method to solve the differential equation

$$ \frac{dx}{dt} = - x^3 + \sin t$$

with initial condition $x=0$ at $t=0$.

#### 1.2.1 Example code

Here is a program to do the calculation from $t=0$ to $t=100$ in 10 steps and plot the results.

```python
# Example code

import numpy as np
import matplotlib.pyplot as plt

def f(x,t):
    return -x**3 + np.sin(t)

a = 0.0           # Start of the interval
b = 10.0          # End of the interval
N = 10            # Number of steps
h = (b-a)/N       # Size of a single step
x = 0.0           # Initial condition

tpoints = np.arange(a,b,h)
xpoints = []
for t in tpoints:
    xpoints.append(x)
    x += h*f(x,t)

plt.plot(tpoints,xpoints)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid()
plt.show()
```

## Part 2: Runge-Kutta method
Euler's method only gives approximate solution neglecting the $\frac{1}{2}h^2 \frac{d^2}{dt^2}$.
To reduce the error, we need to increase the number of steps so that $h$ remains small to keep Euler's method accurate.
Runge-Kutta method is actually a set of methods at different orders, which give results of varying accuracy.

Runge-Kutta method can be derived by performing a Taylor expansion around $t + \frac{1}{2} h$ to get the value of $x(t+h)$:

$$ x(t+h) = x(t+\frac{1}{2}h) + \frac{1}{2}h \left( \frac{dx}{dt}\right)_{t+\frac{1}{2}h}   + \frac{1}{8}h^2 \left( \frac{d^2x}{dt^2}\right)_{t+\frac{1}{2}h} +O(h^3)+\ldots .$$

Similarly, one can derive an expression for $x(t)$:

$$ x(t) = x(t+\frac{1}{2}h) - \frac{1}{2}h \left( \frac{dx}{dt}\right)_{t+\frac{1}{2}h}   + \frac{1}{8}h^2 \left( \frac{d^2x}{dt^2}\right)_{t+\frac{1}{2}h} +O(h^3)$$

Subtracting the second expression from the first, and rearranging the equations gives

$$\begin{align}
x(t+h) &= x(t) + \frac{1}{2}h \left( \frac{dx}{dt}\right)_{t+\frac{1}{2}h}  +O(h^3)+\ldots \\
       &= x(t)  + h f \left(x\left(t+\frac{1}{2}h\right),\frac{1}{2}h\right) +O(h^3)
\end{align}$$

Note that the errors are now $O(h^3)$, smaller than Euler's method.

However, we actually do not know $x\left(t+\frac{1}{2}h\right)$, so we use Euler's method to approximate it to $x(t) + \frac{1}{2}h f\left(t\right)$.

To complete calculation for a single step:

$$\begin{align}
k_1 &= hf(x,t) \\
k_2 &= h f\left(x+\frac{1}{2}k_1,t+\frac{1}{2}h \right) \\
x(t+h) &= x(t)+k_2
\end{align}$$

### 2.1 Runge-Kutta method 2nd-order code example

Let us use the second-order of Runge-Kutta method to solve the same differential equation as we solved using Euler's method with the same condition and $N=10$ steps.

#### 2.1.1 Complete 2nd-order of Runge-Kutta method code
First, we copy-paste the Euler's method code below and make changes to the original code in the single-step update.
Fill in the missing code so that you can make a plot of the solution.

```python
# Fill in the missing code

def f(x,t):
    return -x**3 + np.sin(t)

a = 0.0           # Start of the interval
b = 10.0          # End of the interval
N = 10            # Number of steps
h = (b-a)/N       # Size of a single step
x = 0.0           # Initial condition

tpoints = np.arange(a,b,h)
xpoints = []
for t in tpoints:
    xpoints.append(x)
    k_1 = h*f(x,t)
    k_2 = h*f(x+0.5*k_1,t+ 0.5*h)
    x += k_2  # x(t+h)=x(t)+k_2

plt.plot(tpoints,xpoints)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid()
plt.show()
```

### 2.2 Fourth-order Runge-Kutta method
One can continue the Taylor expansion to include terms of $h^3$ and $h^4$, and so on to cancel out the expressions, and so get a more accurate rule for solving differential equations, carrying an error of order $h^5$.
We will skip the derivation here, but show the final results:

$$\begin{align}
k_1 &= hf(x,t) \\
k_2 &= h f\left(x+\frac{1}{2}k_1,t+\frac{1}{2}h \right) \\
k_3 &= h f\left(x+\frac{1}{2}k_2,t+\frac{1}{2}h \right) \\
k_4 &= h f\left(x+\frac{1}{2}k_3,t+h \right) \\
x(t+h) &= x(t)+\frac{1}{6}( k_1+2k_2+2k_3+k_4)
\end{align}$$

#### 2.2.1 Complete 4th-order of Runge-Kutta method code
Now, you can repeat the steps in 2.1.1 to copy-paste the Euler's method code or your second-order Runge-Kutta method below and make changes to the original code in the single-step update to include $k_3$, $k_4$ and $x$ update.   
Once again, use the exact same conditional as in 1.2.1 to start with $N=10$ steps.

```python
# put your code here

def f(x,t):
    return -x**3 + np.sin(t)

N = np.array([20])         # Number of steps
colors = ['red', 'green', 'blue', 'purple']

for i in range(len(N)):
    h = (b-a)/N[i]       # Size of a single step
    tpoints = np.arange(a,b,h)
    x = 0.0           # Initial condition
    xpoints = []
    for t in tpoints:
        xpoints.append(x)
        k1 = h*f(x,t)
        k2 = h*f(x+0.5*k1, t+0.5*h)
        k3 = h*f(x+0.5*k2, t+0.5*h)
        k4 = h*f(x+k3, t+h)
        x += (1/6)*(k1 + 2*k2 + 2*k3 + k4) # updating the x and y values
        
    plt.plot(tpoints,xpoints, label = 'N = ' + str(N[i]), color = colors[i])
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid()
plt.legend()
plt.show()
```
