## Part 1: Diffraction Limit of a Telescope (15 points)

Our ability to resolve detail in astronomical observations is limited by the diffraction of light in our telescopes.
Light from stars can be treated effectively as coming from a point source at infinity.
When such light with wavelength $\lambda$ passes through the circular aperture of a telescope (which we'll assume to have unit radius) and is focused by the telescope on the focal plane, it produces not a single dot, but a  circular diffraction pattern consisting of central spot surrounded by a series of concentric rings.
The intensity of the light in this diffraction pattern is given by

$$ I(r) = \left( {J_1(kr)\over kr} \right)^2,$$

where $r$ is the distance in the focal plane from the center of the diffraction pattern, $k=2\pi/\lambda$, and $J_1(x)$ is a Bessel function.

The Bessel functions $J_m(x)$ are given by

$$J_m(x) = {1\over\pi} \int_0^\pi d\theta\, \cos(m\theta - x\sin\theta),$$

where $m$ is a nonnegative integer and $x \ge 0$.

### 1.1 Write your own `bessel_J(m, x)` function (5 points)
Write a Python function `bessel_J(m, x)` that calculates the value of $J_m(x)$ using Simpson's rule with $N=1000$ points.
Don't forget to test your function before moving on.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simpson_rule(f, N, a, b):

    if N % 2 != 0:
        raise ValueError("Number of slices (N) must be even for Simpson's rule.")
    
    h = (b - a) / N  # returning the final result
    integral = f(a) + f(b)  # initializing integral with endpoints
    
    # Adding the value of each slice to the integral
    for k in range(1, N): # for ODD numbers
        x = a + k * h
        if k % 2 == 0: # for EVEN numbers
            integral += 2 * f(x)  # Even-indexed terms multiplied by 2
        else:
            integral += 4 * f(x)  # Odd-indexed terms multiplied by 4
    
    integral *= h / 3 # multiplying the sum by the width and divide by 3
    return integral # returning the final result


def bessel_J(m, x):
    n = 1000
    theta = np.linspace(0, np.pi, n)
    integral = simpson_rule(lambda t: np.cos(m*t - x*np.sin(t)), n, 0, np.pi)
    return integral/np.pi
```

## Part 2: Heat Capacity of a Solid (20 points)

Debye's theory of solids gives the heat capacity of a solid at temperature $T$ to be

$$ C_V = 9V\rho k_B \left( \frac{T}{\theta_D} \right)^3 \int_0^{\theta_D/T} dx
      {x^4 e^x\over(e^x-1)^2}, $$
      
where $V$ is the volume of the solid,
      $\rho$ is the number density of atoms,
      $k_B$ is Boltzmann's constant,
      and $\theta_D$ is the so-called _Debye temperature_, a property of solids that depends on their density and speed of sound.

### 2.1 Comment on the given code for Gaussian quadrature (5 points)
You will be using Gaussian quadrature for this problem.
Below is an example code implementing it.
Discuss within your group, and add comments for _every line_ of the code.

```python
# comment the following code

def gaussxw(N):

    # initial approx. to roots of the Legendre polynomial
    a = np.linspace(3, 4*N-1, N)/(4*N+2) # generatin initial approximation array 'a'
    x = np.cos(np.pi*a+1/(8*N*N*np.tan(a))) # computing initial root approximation array 'x'

    # Find roots using Newton's method
    epsilon = 1e-15 # setting a small value for stopping criterion
    delta = 1.0 # initializing delta 
    while delta > epsilon: # starting the loop
        p0 = np.ones(N, float) # initializing array p0 with ones
        p1 = np.copy(x) # initializing array p1 with the current approximation of roots 'x'
        for k in range(1, N): # iterating through each polynomial degree
            p0,p1 = p1,((2*k+1)*x*p1 - k*p0)/(k+1) # computing Legendre polynomials p0 and p1
        dp = (N+1)*(p0-x*p1)/(1-x*x) # computing deriv of the Legendre polynomial
        dx = p1/dp # computing change in root approximation 'x'
        x -= dx # updating root approximation 'x'
        delta = max(np.abs(dx)) # updating delta with max. absolute change in 'x'

    # calculating the weights
    w = 2*(N+1)*(N+1)/(N*N*(1-x*x)*dp*dp)
    return x,w # returning the roots and weights


def gaussxwab(N, a, b):
    x, w = gaussxw(N) # calculating roots and weights using Gaussian quadrature
    return 0.5*(b-a)*x + 0.5*(b+a), 0.5*(b-a)*w  # returning the transformed roots and weights
```

#### 3.1.1 (1 points)
A single cycle of a square wave with amplitude 1.

```python
#this is code that I had already written in PHY 415!

# computing the Fourier coefficients
def compute_cn(v, n, T, f0):
    t = np.linspace(0, T, num=1000)  # time values for integration
    integrand = v(t) * np.exp(-1j * 2 * np.pi * n * f0 * t) # calculating the integrand
    cn = 1/T * np.trapz(integrand, t) # calculating cn using trapezoidal rule
    return cn

# Fourier series expansion using complex form
def complex_fourier_series_expansion(v, t, T, N):
    f0 = 1 / T # fundamental frequency 
    series = np.zeros_like(t, dtype=complex) # initializing the series as an array of zeros with complex dtype

    for n in range(-N, N + 1):
        cn = compute_cn(v, n, T, f0) # computing complex coefficient c_n
        series += cn * np.exp(1j * 2 * np.pi * n * f0 * t) # adding term corresponding to c_n to the series
    return series.real # returning the real part of the series
```

```python
from scipy.signal import square #importing the square wave signal

# defining the square wave signal
T = 2 * np.pi  # period
def v(t):
    return square(t)

# testing the code
t = np.linspace(0, 4 * np.pi, num=1000)
N_values = [1, 3, 5, 7, 9, 11]  # Values of N for testing

# creating subplots for each value of N
fig, axs = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle('Approximation with Different N Values', fontsize=24)

for i, N in enumerate(N_values):
    row = i // 3
    col = i % 3

    approx_signal = complex_fourier_series_expansion(v, t, T, N)
    axs[row, col].plot(t, v(t), label='Original Signal', color='m')
    axs[row, col].plot(t, approx_signal, label=f'Approximation (N={N})', color='b')
    axs[row, col].set_title(f'Approximation with N={N}', fontsize =20)
    axs[row, col].grid()
    
    # Set larger font size for x and y labels
    axs[row, col].set_xlabel('Time', fontsize=20)
    axs[row, col].set_ylabel('Amplitude', fontsize=20)

for ax in axs.ravel():
    ax.legend()

plt.tight_layout(rect=[0, 0, 1, 0.95])
```

```python
# computing the Fourier coefficients using FFT
def compute_fft_coefficients(v, T, N):
    t = np.linspace(0, T, num=N, endpoint=False)  # generating an array of time values
    signal = v(t)  # evaluting the signal at the time values
    coefficients = np.fft.fft(signal)  # computing the FFT of the signal
    return coefficients  # returning the Fourier coefficients

# number of points
N = 1000

# computing the Fourier coefficients
coefficients_square = compute_fft_coefficients(v, T, N)

# plotting the magnitudes of Fourier coefficients for square wave
plt.figure(figsize=(10, 6))
plt.plot(np.abs(coefficients_square), label='Square Wave')
plt.title('Magnitudes of Fourier Coefficients (Square Wave)')
plt.xlabel('k')
plt.ylabel('|c_k|')
plt.grid(True)
plt.show()
```

## Part 4:  The Lotka-Volterra Equations (15 points)

The Lotka-Volterra equations are a mathematical model of predator-prey interactions between biological species.
Let two variables $x$ and $y$ be proportional to the size of the populations of two species, traditionally called "rabbits" (the prey) and "foxes" (the predators).
You could think of $x$ and $y$ as being the population in thousands, say, so that $x=2$ means there are 2000 rabbits.
Strictly, the only allowed values of $x$ and $y$ would then be multiples of 0.001, since you can only have whole numbers of rabbits or foxes.
But 0.001 is a pretty close spacing of values, so it's a decent approximation to treat $x$ and $y$ as continuous real numbers so long as neither gets very close to zero.

In the Lotka-Volterra model the rabbits reproduce at a rate proportional to their population, but are eaten by the foxes at a rate proportional to both their own population and the population of foxes:
$$
\frac{d x}{d t} = \alpha x - \beta xy,
$$

where $\alpha$ and $\beta$ are constants.
At the same time, the foxes reproduce at a rate proportional the rate at which they eat rabbits, because they need food to grow and reproduce. They also die of old age at a rate proportional to their own population:

$$
\frac{d y}{d t} = \gamma xy - \delta y,
$$
where $\gamma$ and $\delta$ are also constants.

### 4.1: Solve with Runge-Kutta method (5 points)
Write a program to solve these equations using the fourth-order Runge-Kutta method for the case $\alpha=1$, $\beta = \gamma = 0.5$, and $\delta=2$, starting from the initial condition $x=y=2$.

```python
def f(x, y, t):
    # parameters
    alpha = 1
    beta = 0.5
    gamma = 0.5
    delta = 2
    # derivatives
    dxdt = alpha*x - beta*x*y
    dydt = gamma*x*y - delta*y
    return dxdt, dydt #returning the derivatives

a = 0.0
b = 20.0
N = 10000 # Number of steps
h = (b-a)/N # size of a single step
tpoints = np.arange(a, b, h)
x = 2.0 # initial condition for x
y = 2.0 # intial condition for y
xpoints = []
ypoints = []

for t in tpoints:
    xpoints.append(x)
    ypoints.append(y)
    dxdt, dydt = f(x, y, t)
    k1x, k1y = h*dxdt, h*dydt
    dxdt, dydt = f(x+0.5*k1x, y+0.5*k1y, t+0.5*h)
    k2x, k2y = h*dxdt, h*dydt
    dxdt, dydt = f(x+0.5*k2x, y+0.5*k2y, t+0.5*h)
    k3x, k3y = h*dxdt, h*dydt
    dxdt, dydt = f(x+k3x, y+k3y, t+h)
    k4x, k4y = h*dxdt, h*dydt
    x += (1/6)*(k1x + 2*k2x + 2*k3x + k4x) # updating the x value
    y += (1/6)*(k1y + 2*k2y + 2*k3y + k4y) # updating the y value
```
