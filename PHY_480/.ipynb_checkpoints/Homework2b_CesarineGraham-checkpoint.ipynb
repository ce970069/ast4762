```python
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
import numpy as np
from skimage import measure

# setting up mesh
n = 100 
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)
z = np.linspace(-2,2,100)
X, Y, Z =  np.meshgrid(x, y, z)

def f_heart(x,y,z):
    F = ((x**2 + (9/4)*y**2 + z**2 - 1)**3 - x**2 * z**3 - (9/200)*y**2 * z**3)
    return F

# obtaining value to at every point in mesh
vol = f_heart(X,Y,Z) 

# Extract a 2D surface mesh from a 3D volume
verts, faces, normals, values = measure.marching_cubes(vol, 0,  spacing=(0.1, 0.1, 0.1))

# creating the 3D figure
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# plotting the surface
surf = ax.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2], cmap='RdPu', lw=1)

# setting the initial view and plotting
ax.view_init(elev=15, azim=15)
ax.set_facecolor("pink")
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
import numpy as np
from skimage import measure

# setting up the mesh
n = 100 
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)
z = np.linspace(-2,2,100)
X, Y, Z =  np.meshgrid(x, y, z)

def f_heart(x,y,z):
    F = ((x**2 + (9/4)*y**2 + z**2 - 1)**3 - x**2 * z**3 - (9/200)*y**2 * z**3)
    return F

# obtaining value to at every point in mesh
vol = f_heart(X,Y,Z) 

# extracting the 2D surface mesh from a 3D volume (F=0)
verts, faces, normals, values = measure.marching_cubes(vol, 0,  spacing=(0.1, 0.1, 0.1))

# creating the 3D figure
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

# plotting the surface
surf = ax.plot_trisurf(verts[:, 0], verts[:,1], faces, verts[:, 2], cmap='RdPu', lw=1)
ax.set_facecolor("pink")

# function to update the plot for each frame
def update(frame, ax, surf):
    ax.view_init(15, frame)  # Adjust the view for rotation
    return surf,

# number of frames for the animation
num_frames = 360

# creating and saving the animation as a gif
ani = FuncAnimation(fig, update, frames=num_frames, fargs=(ax, surf), interval=50, blit=True)
ani.save('heart_rotation.gif', writer='imagemagick', fps=30)
plt.show()
```
