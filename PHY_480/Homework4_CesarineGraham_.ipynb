## Part 1: Electronic Capacitor (20 points)

Consider the following simple model of an electronic capacitor, consisting of two flat metal plates enclosed in a square metal box, as shown in [this picture](https://drive.google.com/file/d/1Vsvx1QRS7-vDF_3FtT6z41yhnod2FAKf/view?usp=sharing).
For simplicity, let us model the system in two dimensions.
Using any of the methods we have studied, write a program to calculate the electrostatic potential in the box on a grid of $100\times 100$ points, where the walls of the box are at voltage zero and the two plates (which are of negligible thickness) are at voltages $\pm 1\text{ V}$ as shown.

_Hint: Notice that the capacitor plates are at fixed voltage, not fixed charge, so this problem differs from the problem with the two charges in the Day15 IC notebooks. In effect, the capacitor plates are part of the boundary condition in this case: they behave the same way as the walls of the box, with potentials that are fixed at a certain value and cannot change._


### 1.1 Write the code (10 points)
Write a program to calculate the value of the potential at each grid point to a precision of
$10^{-6}\text{ V}$.

```python
from numpy import empty, zeros, max, copy
from pylab import imshow, viridis, show, colorbar, title

# Constants
M = 100         # Grid squares on a side
V_plate = 1.0   # Voltage of the plates
V_wall = 0.0    # Voltage of the walls
target = 1e-6   # Target accuracy

# Create arrays to hold potential values
phi = zeros([M+1, M+1], float)
phi[:, 0] = V_wall
phi[:, -1] = V_wall
phi[25:75, 25] = V_plate
phi[25:75, 75] = -V_plate
phinew = copy(phi)

# Main loop
delta = 1.0
while delta > target:
    # Calculate new values of the potential
    for i in range(1, M):
        for j in range(1, M):
            if 25 <= i <= 74 and (j == 25 or j == 75):
                continue  # Skip the plates
            phinew[i, j] = (phi[i+1, j] + phi[i-1, j] + phi[i, j+1] + phi[i, j-1])/4
    # Calculate maximum difference from old values
    delta = max(abs(phi - phinew))
    # Swap the two arrays around
    phi, phinew = phinew, phi
```

Make a density plot of the result.

```python
# plotting
imshow(phi, origin='lower')
viridis()
colorbar()
title('Potential')
show()
```

## Part 3: Brownian Motion (20 points)

Brownian motion is the motion of a particle, such as a smoke or pollen, in a fluid, as it is buffeted by random collisions with the surrounding molecules.
Make a simple computer simulation of such a particle in two dimensions as follows.  

The particle is confined to a square grid or lattice $L\times L$ squares on a side, so that its position can be represented by two integers $i,j \in [0, L-1]$.
It starts in the middle of the grid and undergoes a random walk.
On each step of the simulation, choose a random direction (up, down, left, or right) and move the particle one step in that direction.
The particle is not allowed to move outside the limits of the lattice; if it tries to do so, choose a new random direction to move in.

### 3.1 Write code to per
Brownian motion in 2D (10 points)
Write a program to perform a million steps of this process on a lattice with $L=101$.
(We chose an odd length for the side of the square so that there is one lattice site exactly in the center.)

### 3.2 Visualize your result (10 points)
Make an animation on the screen of the position of the particle.
If you need an example of how to make an animation, please check the end of this notebook.

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output, display
import time

def step2d():
    direction = np.random.randint(0, 4)  # 0: left, 1: right, 2: up, 3: down
    if direction == 0:
        return (-1, 0)  # left
    elif direction == 1:
        return (1, 0)   # right
    elif direction == 2:
        return (0, 1)   # up
    else:
        return (0, -1)  # down

Nsteps = 10000  # Number of steps
x, y = 0, 0    # Initial position at the origin
heatmap = np.zeros((101, 101))  # Grid for the heatmap, assuming a 101x101 grid
x_hold, y_hold = [], []

fig, ax = plt.subplots(figsize=(8, 8))

for i in range(Nsteps):
    dx, dy = step2d()
    x += dx
    y += dy
    x_hold.append(x)
    y_hold.append(y)
    heatmap[y+50, x+50] += 1  # Update the heatmap grid
    ax.clear()
    ax.imshow(heatmap, cmap='hot', interpolation='nearest')
    ax.plot(x_hold, y_hold, 'bo-', markersize=1, alpha=0.5)
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.set_aspect('equal', adjustable='box')
    ax.set_title('2D Random Walk with Heatmap', pad=20)
    ax.text(0.5, 1.01, "The brighter the color, the more frequent the particle has visited that particular grid point",
            horizontalalignment='center', verticalalignment='bottom', transform=ax.transAxes)
    clear_output(wait=True)
    display(fig)
    time.sleep(0.0001)

plt.close()
```
