### Exercise 1 (10pt), Equations of motion and more

The equations of motion in the center-of-mass frame in two dimensions with $x=r\cos{(\phi)}$ and $y=r\sin{(\phi)}$ and
$r\in [0,\infty)$, $\phi\in [0,2\pi]$ and $r=\sqrt{x^2+y^2}$ are given by

$$
\mu \ddot{r}=-\frac{dV(r)}{dr}+\mu r\dot{\phi}^2,
$$

and

$$
\dot{\phi}=\frac{L}{\mu r^2}.
$$

Here $V(r)$ is any central force which depends only on the relative coordinate.
* 1a (5pt) Show that you can rewrite the radial equation in terms of an effective potential $V_{\mathrm{eff}}(r)=V(r)+L^2/(2\mu r^2)$. 

* 1b (5pt) Write out the final differential equation for the radial degrees of freedom when we specify that $V(r)=-\alpha/r$.  Plot the effective potential. You can choose values for $\alpha$ and $L$ and discuss (see Taylor section 8.4 and example 8.2) the physics of the system for two energies, one larger than zero and one smaller than zero. This is similar to what you did in the first midterm, except that the potential is different.

<span style="color:darkorchid">
1a.
\begin{align*} 
E &= K + V \\ 
 &= \frac{1}{2} \mu (\dot{r}^{2} + r^{2} \dot{\phi}^{2}) + V(r)  \\
 &= \frac{1}{2} \mu \dot{r}^{2} + V_{\mathrm{eff}} \\
V_{\mathrm{eff}} &= V(r) + \frac{L^{2}}{2 \mu r^{2}}
\end{align*}

```python
#1b: Plotting the Potential

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

deltaR = 0.0001
a = 1.0
L = 1.0
mu = 1

r0 = 1.0
rf = 2.0
n = int(np.ceil((rf-r0)/deltaR))
r = np.zeros(n)

for i in range(n):
    r[i] = r0+i*deltaR
U = np.zeros(n)
U = -(a/r**2)+ (L**2/(mu*r**3))

plt.plot(r, U, color = "darkorchid", label = 'potential')
plt.title("1a : Plotting the Potential")
plt.xlabel("r(x,y) [distance]")
plt.ylabel("U(r) [potential energy]")
plt.tight_layout()
plt.legend()
plt.grid()
plt.show()
```

Make a class called Solvers which contains the following:
1. An initializer

2. A method that implements Euler's method in 2D

3. A method that implements Euler-Cromer's method in 2D

The class may also contain any other methods or class level variables that you think are needed.

Make then a function (outside of the class Solver) that solves the Earth-Sun problem using the Solvers class (Exercise 6 from Homework 5).

```python
import numpy as np
import matplotlib.pyplot as plt

class Solvers:
    def __init__(self, initial_position, initial_velocity, acceleration_func):
        self.position = initial_position
        self.velocity = initial_velocity
        self.acceleration_func = acceleration_func
    
    def euler_method_2d(self, dt, num_steps):
        positions = [self.position]
        velocities = [self.velocity]
        
        for i in range(num_steps):
            acceleration = self.acceleration_func(positions[-1])
            new_velocity = velocities[-1] + acceleration * dt
            new_position = positions[-1] + new_velocity * dt
            velocities.append(new_velocity)
            positions.append(new_position)
        
        self.position = positions[-1]
        self.velocity = velocities[-1]
        return np.array(positions)
    
    def euler_cromer_method_2d(self, dt, num_steps):
        positions = [self.position]
        velocities = [self.velocity]
        
        for i in range(num_steps):
            acceleration = self.acceleration_func(positions[-1])
            new_velocity = velocities[-1] + acceleration * dt
            new_position = positions[-1] + new_velocity * dt
            positions.append(new_position)
            velocities.append(new_velocity)
        
        self.position = positions[-1]
        self.velocity = velocities[-1]
        return np.array(positions)
    
    def velocity_verlet_2d(self, dt, num_steps):
        positions = [self.position]
        velocities = [self.velocity]
        
        for i in range(num_steps):
            acceleration = self.acceleration_func(positions[-1])
            new_position = positions[-1] + velocities[-1] * dt + 0.5 * acceleration * dt ** 2
            new_acceleration = self.acceleration_func(new_position)
            new_velocity = velocities[-1] + 0.5 * (acceleration + new_acceleration) * dt
            positions.append(new_position)
            velocities.append(new_velocity)
            acceleration = new_acceleration
        
        self.position = positions[-1]
        self.velocity = velocities[-1]
        return np.array(positions)
    
    def plot_results(self, positions, dt):
        times = np.arange(0, len(positions)*dt, dt)
        plt.plot(times, positions[:, 0], label='x')
        plt.plot(times, positions[:, 1], label='y')
        plt.xlabel('Time (s)')
        plt.ylabel('Position (m)')
        plt.legend()
        plt.show()

def earth_sun_solver():
    # Define gravitational constant
    G = 6.67430e-11
    
    # Define initial conditions
    earth_position = np.array([1.4710e11, 0])
    earth_velocity = np.array([0, 3.0287e4])
    sun_position = np.array([0, 0])
    sun_mass = 1.989e30
    
    def earth_acceleration(position):
        r = np.linalg.norm(position - sun_position)
        return -G * sun_mass * (position - sun_position) / r**3
    
    # Initialize solver
    solver = Solvers(earth_position, earth_velocity, earth_acceleration)
    
    # Solve with Velocity-Verlet method
    dt = 1000
    num_steps = 365
    positions = solver.velocity_verlet_2d(dt, num_steps)
    
    # Plot results
    solver.plot_results(positions, dt)

# Run the earth_sun_solver function
earth_sun_solver()
```
