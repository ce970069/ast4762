### Exercise 3 (10 pt), The Lennard-Jones potential

[The Lennard-Jones potential](https://en.wikipedia.org/wiki/Lennard-Jones_potential) is often used to describe
the interaction between two atoms or ions or molecules. If you end up doing materals science and molecular dynamics calculations, it is very likely that you will encounter this potential model.
The expression for the potential energy is
of the molecule is:

$$
V(r) = V_0\left((\frac{a}{r})^{12}-(\frac{b}{r})^{6}\right),
$$

where $V_0$, $a$ and $b$ are constants and the potential depends only on the relative distance between two objects
$i$ and $j$, that is $r=\vert\boldsymbol{r}_i-\boldsymbol{r}_j\vert=\sqrt{(x_i-x_j)^2+(y_i-y_j)^2+(z_i-z_j)^2}$.

* 3a (3pt) Sketch/plot the potential (choose some values for the constants in doing so).

* 3b (3pt) Find and classify the equilibrium point(s).

* 3c (4pt) What is the force acting on one of the objects (an atom for example) from the other object? Is this a conservative force?

```python
#3a. Plotting the Potential

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

deltaX = 0.0001
V0 = 5.0
a = 1.0
b = 1.0

x0 = 1.0
xf = 2.0
n = int(np.ceil((xf-x0)/deltaX))
x = np.zeros(n)

for i in range(n):
    x[i] = x0+i*deltaX
U = np.zeros(n)
U = V0*( ((a/x)**12) - ((b/x)**6) )

plt.plot(x, U, color = "darkviolet", label = 'potential')
plt.title("3a : Plotting the Potential")
plt.xlabel("x(t) [distance]")
plt.ylabel("U(x) [potential energy]")
plt.tight_layout()
plt.legend()
plt.grid()
plt.show()
```

```python
#3b. Finding and Classifying the Equilibrium Points

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

deltaX = 0.0001
V0 = 5.0
a = 1.0
b = 1.0

x0 = 1.0
xf = 2.0
n = int(np.ceil((xf-x0)/deltaX))
x = np.zeros(n)

for i in range(n):
    x[i] = x0+i*deltaX
U = np.zeros(n)
U = V0*( ((a/x)**12) - ((b/x)**6) )
U_deriv = V0*( - ((12*a**12/x**13)) + ((6*b**6/x**7)) )

plt.plot(x, U, color = "darkviolet", label = 'potential')
plt.plot(x, U_deriv, color = "mediumvioletred", label = 'potential derivative')
plt.title("3b : Finding and Classifying the Equilibrium Points")
plt.xlabel("x(t) [distance]")
plt.ylabel("U(x) [potential energy]")
plt.ylim(-1.3,0)
plt.legend()
plt.grid()
```

<span style="color:darkviolet"> At the bottom of the potential well, the derivative of the potential is equal to 0. This means that this is a stable point and there are no external forces acting on the system.

<span style="color:darkviolet"> **3c**. Since there is a potential well which has no external forces acting upon it, this forces is conservative. An example of a force that this could be is an electromagnetic force, which typically create wells such as the one plotted above.

```python
#4C.

#Part 7
#Using the Euler-Cromer method in order to conserve energy of the system

#needed imports
import numpy as np
import pandas as pd
from math import *
import matplotlib.pyplot as plt

#initial conditions
k = 1
m = 1
b = 1
w0 = (k/m)**(1/2)
gamma = b/(2*m*w0*w0) #for underdamping
beta = b/(2*m)
w2 = np.sqrt(w0**2-2*beta**2)
a1 = 1
a2 = 1

#setting up arrays 
DeltaT = 0.001
tf = 20
n = int(np.ceil(tf/DeltaT))
t = np.zeros(n)
v = np.zeros(n)
x = np.zeros(n)
v0 = 1
x0 = 1
x[0] = x0
v[0] = x0

#integrating using Euler-Cromer's method
for i in range(n-1):
    a = -gamma*v[i]-w0**2*x[i]
    
    #updating velocity, time and position arrays
    v[i+1] = v[i] + DeltaT*a
    x[i+1] = x[i] + DeltaT*(v[i+1]) + DeltaT*DeltaT*a
    t[i+1] = t[i] + DeltaT
    
#plotting position as function of time    
plt.figure(figsize=(12,10))
plt.subplot(2,1,1)
plt.plot(t, x, color = "red", label = 'x[m]')
plt.xlim([0,tf])
plt.title('position over time for underdamping')
plt.xlabel('time [s]')
plt.ylabel('position [m]')
plt.grid()
plt.legend()

#plotting velocity as function of time    
plt.subplot(2,1,2)
plt.plot(t, v, color = 'black', label = 'v[m/s]')
plt.xlim([0,tf])
plt.title('velocity over time for underdamping')
plt.xlabel('time [s]')
plt.ylabel('velocity [m/s]')
plt.hlines(0,0,20, color = 'green')
plt.grid()
plt.legend()
```

```python
#4C.

#Part 7
#Using the Euler-Cromer method in order to conserve energy of the system

#needed imports
import numpy as np
import pandas as pd
from math import *
import matplotlib.pyplot as plt

#initial conditions
k = 1
m = 1
b = 1
w0 = (k/m)**(1/2)
gamma = 2 #overdamping
beta = b/(2*m)
w2 = np.sqrt(w0**2-2*beta**2)
a1 = 1
a2 = 1

#setting up arrays 
DeltaT = 0.001
tf = 20
n = int(np.ceil(tf/DeltaT))
t = np.zeros(n)
v = np.zeros(n)
x = np.zeros(n)
v0 = 1
x0 = 1
x[0] = x0
v[0] = x0

#integrating using Euler-Cromer's method
for i in range(n-1):
    a = -gamma*v[i]-w0**2*x[i]
    
    #updating velocity, time and position arrays
    v[i+1] = v[i] + DeltaT*a
    x[i+1] = x[i] + DeltaT*(v[i+1]) + DeltaT*DeltaT*a
    t[i+1] = t[i] + DeltaT
    
#plotting position as function of time    
plt.figure(figsize=(12,10))
plt.subplot(2,1,1)
plt.plot(t, x, color = "red", label = 'x[m]')
plt.xlim([0,tf])
plt.title('position over time for overdamping')
plt.xlabel('time [s]')
plt.ylabel('position [m]')
plt.grid()
plt.legend()

#plotting velocity as function of time    
plt.subplot(2,1,2)
plt.plot(t, v, color = 'black', label = 'v[m/s]')
plt.xlim([0,tf])
plt.title('velocity over time for overdamping')
plt.xlabel('time [s]')
plt.ylabel('velocity [m/s]')
plt.hlines(0,0,20, color = 'green')
plt.grid()
plt.legend()
```
