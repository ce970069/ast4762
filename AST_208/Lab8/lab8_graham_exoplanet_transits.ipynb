#### Conversions and Formulas for TESS C Host Star:
    
Radius ($R_J$) : $16.0483 \pm 0.5675$
        
$\Delta m =  \frac{ \pi r_{p}^{2} }{ \pi r_{*}^{2} }   $  
$  r_{p} = r_{*}  \sqrt{ \Delta m}   $

<span style="color:seagreen">
    
#### Radius of the Planet:

\begin{align*}
r_{p} &= r_{*}  \sqrt{ \Delta m} \\  
r_{p} &= 16.0483 \sqrt{0.008} \\
r_{p} &= 1.435 R_{J}
\end{align*}

<span style="color:seagreen">
    
#### Uncertainty Propagation of the Radius of the Planet: 
    
\begin{align*}
\Delta  r_{p}  &= \sqrt{\left( \frac{\partial r_{p}}{\partial r_{\odot}} \delta r_{\odot} \right)^2 + \left( \frac{\partial r_{p}}{\partial m} \delta (\Delta m) \right)^2} \\
\Delta  r_{p}  &= \sqrt{\left(  \sqrt{m} \cdot \delta r_{\odot} \right)^2 + \left( \frac{-\ r_{\odot} \sqrt{m} }{2} \delta (\Delta m ) \right)^2} \\
\Delta  r_{p}  &= 0.051545453 \\
\Delta  r_{p}  &= 5.1545453 * 10^{-2} \\
\end{align*} 

#### Radius of the Planet with Uncertainty:
\begin{align*}
r_{p} = 1.435\pm 0.05 R_{J}
\end{align*}

<i class="fa fa-cogs" style="font-size:1.5em; color:red"></i> EC) Estimate the surface temperatures for your planet, and their uncertainties. If the atmosphere is neglected, the surface temperature can be approximated as
\begin{equation}
 T_p = T_\star \left(\frac{R_\star}{2d}\right)^{1/2} \left(1-A\right)^{1/4} .
\end{equation}
Here $T_\star$ is the stellar surface temperature, $R_\star$ is the stellar radius, $d$ is the star-planet distance, and $A$ is the planet's *albedo*the fraction of light reflected from the planet, so that clouds have $A\approx 1$ and black lava has $A\ll 1$. We have no clue what $A$ is, but for a first guess let's assume it is equal to that of Earth, $A\approx0.4$.


<span style="color:seagreen">

#### Surface Temperature for Planet C:

\begin{align*}
T_p &= T_\star \left(\frac{R_\star}{2d}\right)^{1/2} \left(1-A\right)^{1/4} \\
T_p &= (6630) \left(\frac{1.612}{2(10.8892)}\right)^{1/2} \left(1-0.4\right)^{1/4} \\
T_p &= 1587.5 K
\end{align*}

<span style="color:seagreen">
   
#### Uncertainty Propagation of the Surface Temperature of Planet C: 

\begin{align*}
\Delta{Tp} &= \sqrt{\left(\frac{\partial Tp}{\partial Ts}\delta{Ts}\right)^2 + \left(\frac{\partial Tp}{\partial Rs}\delta{Rs}\right)^2 + \left(\frac{\partial Tp}{\partial d}\delta{d}\right)^2 + \left(\frac{\partial Tp}{\partial A}\delta{A}\right)^2} \\
\frac{\partial Tp}{\partial Ts} &= \left(\frac{Rs}{2d}\right)^{\frac{1}{2}} (1-A)^{\frac{1}{4}} \\
\frac{\partial Tp}{\partial Rs} &= \frac{Ts}{2d}\left(\frac{Rs}{2d}\right)^{\frac{1}{2}} (1-A)^{\frac{1}{4}} \\
\frac{\partial Tp}{\partial d} &= -\left(\frac{TsRs}{4d^2}\right)^{\frac{1}{2}} (1-A)^{\frac{1}{4}} \\
\frac{\partial Tp}{\partial A} &= -\frac{1}{4}\left(\frac{TsRs}{2d}\right)^{\frac{1}{2}} (1-A)^{-\frac{3}{4}} \\
\Delta{Tp} &= 100 K
\end{align*}
    
$\therefore T_{p} = 1587.5 \pm 100 K$
