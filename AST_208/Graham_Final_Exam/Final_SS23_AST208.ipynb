<i class="fa fa-pencil" style="font-size:1.5em; color:red"></i> 3 e) Estimate the maximum line-of-sight velocity $K$ of the star? Explain your reasoning. (2 pts)

```python
#load data in from text file
data = np.genfromtxt('Final_RVs_SS23.dat')

#arrays for each column
JD = data[:,0]
RV = data[:,1]
RV_err = data[:,2]

#making a time array
t = JD - JD[0]

#mkaing the scatter plot with error bars
plt.errorbar(t,RV,yerr=RV_err,linestyle='none',marker='o', color='seagreen')
plt.xlabel("Time [Days]")
plt.ylabel("Radial Velocity [m/s]")
plt.title("Radial Velocity over Time")
plt.grid()
```

```python
t_fold = np.remainder(t,9.4832) / 9.4832
plt.errorbar(t_fold, RV,yerr=RV_err,linestyle='none',marker='o', color='seagreen')
plt.xlabel("Phase [$\phi$]")
plt.ylabel("Radial Velocity [m/s]")
plt.title("Radial Velocity Curve of TESS")
plt.grid()
```

```python
phi = np.arange(0, 1, 0.0001)

#turning a sine curve fit into a function
def fit(phi, v0, K, phi_0):
    return v0 + K*np.sin((2*np.pi)*(phi+phi_0))

para,cov = curve_fit(fit,t_fold,RV)
RV_fit = fit(phi,para[0],para[1],para[2])

plt.xlabel("Phase [$\phi$]")
plt.ylabel("Radial Velocity [m/s]")
plt.title("Radial Velocity Curve of TESS")
plt.grid()
plt.plot(phi, RV_fit, color = 'm')

plt.errorbar(t_fold, RV,yerr=RV_err,linestyle='none',marker='o', color='seagreen')
```

<span style="color:seagreen">
I estimate the value of the maximum line-of-sight velocity $K$ of the star to be -428.88616862495616 with the above fit

```python
G = scipy.constants.G
M_s = 1.777 * 1.989e30
P = 9.4832 * 24 * 3600
K = abs(para[1])

Mp = K /(((2*np.pi*G)/(P*(M_s**2)))**(1/3))
M_p = Mp * 1.899e-27 
print("the mass of the planet is ", M_p, "Jupiter mass.")
```
