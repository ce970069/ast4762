## Where do we start? Compartmental models

The modeling of epidemics follows from a few simple considerations:

* the population under consideration ca be sufficiently compartmentalized into distinct groups, and
* there is a rate (fixed or variable) at which individuals can move between compartments.

The [compartmental model](https://en.wikipedia.org/wiki/Compartmental_models_in_epidemiology) is a simple way to model the spread of a disease and where we will start.

### The SIS model

The Susceptible-Infected-Susceptible (SIS) model describes an epidemic in which healthy individuals (Susceptibles) can become infected with the disease in question (Infected). The infected individuals can be cured but retain no natural immunity to the disease (i.e., become susceptible again). The figure below illustrates the compartments and flux of individuals in the SIS model. Such models are suitable for bacterial infections.

![The SIS model](../assets/images/sis.png)

The dynamics are given by,

$$\dfrac{dS}{dt} = -\beta S I + \gamma I$$
$$\dfrac{dI}{dt} = \beta S I - \gamma I$$

The rate at which susceptibles are infected is related to the contact rate ($\beta$). The recovery rate ($\gamma$) describes how quickly individuals are cured. The rate equations that govern this model are two ordinary non-linear differential equations. Here we have neglected the birth and death rates of the populations by making the approximation the epidemic occurs very quickly. If we assume there's a total number of people $N$, then we can write these equations as:

$$\dfrac{dS}{dt} = -\beta S I + \gamma (N - S)$$
$$I(t) = N - S(t)$$

For given values of $\beta$ and $\gamma$, we can solve these equations numerically.

### The SIR model

The Susceptible-Infected-Recovered (SIR) model describes an epidemic in which healthy individuals (Susceptibles) can become infected with the disease in question (Infected). The infected individuals can be cured and retain a natural immunity to the disease (Recovered). The figure below illustrates the compartments and flux of individuals in the SIR model. Such models are suitable for viral infections. We can modify this model for the zombies.

![The SIR model](../assets/images/sir.png)

The transfer rates can be defined for the SIR model as was for the SIS model. The rate equations that govern this model are three ordinary non-linear differential equations. Again, we have neglected the birth and death rates of the populations by making the approximation the epidemic occurs very quickly.

$$\dfrac{dS}{dt} = -\beta S I$$
$$\dfrac{dI}{dt} = \beta S I - \gamma I$$
$$\dfrac{dR}{dt} = \gamma I$$

Again, for given values of $\beta$ and $\gamma$, we can solve these equations numerically.

### The SZR model

This SZR model treats the population as compartmentalized into the 3 groups: Susceptible (S), Zombie (Z), and Removed (R). Movement between these groups is illustrated below.

![The SZR model](../assets/images/szr.png)

Susceptibles can become zombies through an encounter with a zombie. Zombies can move to the removed compartment by being destroyed in classic manners (e.g., Dawn of the Dead). Susceptibles can move to the removed compartment through death by a non-zombie encounter. Finally, removed individuals can become zombies through typical resurrection techniques (e.g., Live and Let Die).


The deterministic model includes transfer rates we have seen before, but we include the rates of transfer from susceptible to removed ($\delta$) and removed to zombie ($\xi$).

$$\dfrac{dS}{dt} = - \beta S Z - \delta S$$
$$\dfrac{dZ}{dt} = \beta S Z + \xi R - \gamma S Z$$
$$\dfrac{dR}{dt} = \delta S +\gamma SZ - \xi R$$

For given values of values of $\beta$, $\gamma$, $\delta$, and $\xi$, we can solve these equations numerically.

## SIS Model

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# SIS Model
def SIS(time, current_state, beta, gamma):
    
    susceptible, infected = current_state
    
    dsusceptible_dt = -beta * susceptible * infected + gamma * infected
    dinfected_dt = beta * susceptible * infected - gamma * infected
    
    return [dsusceptible_dt, dinfected_dt]

beta_sis = 0.002
gamma_sis = 0.1

initial_conditions_sis = [100, 50]  # Initial values for S, I
time_sis = np.linspace(0, 200, 1000)  # Adjust the time range as needed

result_sis = solve_ivp(SIS, (0, 200), initial_conditions_sis, t_eval=time_sis, args=(beta_sis, gamma_sis))

susceptible_sis = result_sis.y[0, :]
infected_sis = result_sis.y[1, :]

plt.plot(time_sis, susceptible_sis, label='susceptible', color='limegreen')
plt.plot(time_sis, infected_sis, label='infected', color='firebrick')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('population (thousands)')
plt.title('Solution of the SIS Model')
plt.show()
```

## SIR Model

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# SIR Model
def SIR(time, current_state, beta, gamma):
    
    susceptible, infectious, recovered, asymptomatic, drug_before_symptoms, drug_during_symptoms = current_state
    
    dsusceptible_dt = -beta * susceptible * infectious
    dinfectious_dt = beta * susceptible * infectious - gamma * infectious
    drecovered_dt = gamma * infectious
    dasymptomatic_dt = beta * susceptible * asymptomatic - gamma * asymptomatic
    drug_before_dt = beta * susceptible * drug_before_symptoms - gamma * drug_before_symptoms
    drug_during_sympoms_dt = beta * susceptible * drug_during_symptoms - gamma * drug_during_symptoms
    
    return [dsusceptible_dt, dinfectious_dt, drecovered_dt, dasymptomatic_dt, drug_before_dt, drug_during_sympoms_dt]

beta = 0.002
gamma = 1

initial_conditions = [1000, 1, 0, 3, 2, 4]
time = np.linspace(0, 20, 1000)

result = solve_ivp(SIR, (0, 20), initial_conditions, t_eval=time, args=(beta, gamma))

susceptible = result.y[0, :]
infectious = result.y[1, :]
recovered = result.y[2, :]
asymptomatic = result.y[3, :]
drug_before_symptoms = result.y[4, :]
drug_during_symptoms = result.y[5, :]

plt.plot(time, susceptible, label='susceptible')
plt.plot(time, infectious, label='infectious')
plt.plot(time, recovered, label='recovered')
plt.plot(time, asymptomatic, label='asymptomatic')
plt.plot(time, drug_before_symptoms, label='drugs before symptoms')
plt.plot(time, drug_during_symptoms, label='drugs during symptoms')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('population (thousands)')
plt.title('Solution of the SIR Model')
plt.show()
```

## SZR Model

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# SZR Model
def SZR(time, current_state, beta, gamma, delta, xi):
    
    susceptible, zombie, removed = current_state
    
    dsusceptible_dt = -beta * susceptible * zombie - delta * susceptible
    dzombie_dt = beta * susceptible * zombie + xi * removed - gamma * susceptible * zombie
    dremoved_dt = delta * susceptible + gamma * susceptible * zombie - xi * removed
    
    return [dsusceptible_dt, dzombie_dt, dremoved_dt]

beta = 0.002
gamma = 0.1
delta = 0.1
xi = 0.05

initial_conditions_szr = [1000, 1, 0]  # Initial values for S, Z, R
time_szr = np.linspace(0, 200, 1000)  # Adjust the time range as needed

result_szr = solve_ivp(SZR, (0, 200), initial_conditions_szr, t_eval=time_szr, args=(beta, gamma, delta, xi))

susceptible_szr = result_szr.y[0, :]
zombie_szr = result_szr.y[1, :]
removed_szr = result_szr.y[2, :]

plt.plot(time_szr, susceptible_szr, label='susceptible', color='limegreen')
plt.plot(time_szr, zombie_szr, label='zombie', color='firebrick')
plt.plot(time_szr, removed_szr, label='removed', color='darkorchid')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('population (thousands)')
plt.title('Solution of the SZR Model')
plt.show()
```

## SZQR Model:

Quarantined Susceptible (Q) and Quarantined Zombie (QZ). Individuals in the susceptible compartment can move to the quarantined susceptible compartment, and individuals in the zombie compartment can move to the quarantined zombie compartment. The dynamics of the model can be described by the following set of equations, where $\alpha\$ represents the rate at which individuals move from susceptible and zombie compartments to the corresponding quarantined compartments:

$$
\begin{align*}
\dfrac{dS}{dt} &= -\beta S Z - \delta S - \alpha S Q \\
\dfrac{dZ}{dt} &= \beta S Z + \xi R - \gamma S Z - \alpha Z Q \\
\dfrac{dR}{dt} &= \delta S + \gamma SZ - \xi R + \alpha Q \\
\dfrac{dQ}{dt} &= \alpha S Q + \alpha Z Q \\
\dfrac{dQZ}{dt} &= \alpha Z Q
\end{align*}
$$

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# SZQR Model
def SZQR(time, current_state, beta, gamma, delta, xi, alpha):
    
    susceptible, zombie, removed, quarantined_susceptible, quarantined_zombie = current_state
    
    dsusceptible_dt = -beta * susceptible * zombie - delta * susceptible - alpha * susceptible * quarantined_susceptible
    dzombie_dt = beta * susceptible * zombie + xi * removed - gamma * susceptible * zombie - alpha * zombie * quarantined_zombie
    dremoved_dt = delta * susceptible + gamma * susceptible * zombie - xi * removed + alpha * quarantined_susceptible
    dquarantined_susceptible_dt = alpha * susceptible * quarantined_susceptible
    dquarantined_zombie_dt = alpha * zombie * quarantined_zombie
    
    return [dsusceptible_dt, dzombie_dt, dremoved_dt, dquarantined_susceptible_dt, dquarantined_zombie_dt]

beta_szqr = 0.002
gamma_szqr = 0.1
delta_szqr = 0.1
xi_szqr = 0.05
alpha_szqr = 0.02  # Adjust the quarantine rate as needed

initial_conditions_szqr = [10000, 1, 0, 0, 10]  # Initial values for S, Z, R, Q, QZ
time_szqr = np.linspace(0, 200, 1000)  # Adjust the time range as needed

result_szqr = solve_ivp(SZQR, (0, 200), initial_conditions_szqr, t_eval=time_szqr, args=(beta_szqr, gamma_szqr, delta_szqr, xi_szqr, alpha_szqr))

susceptible_szqr = result_szqr.y[0, :]
zombie_szqr = result_szqr.y[1, :]
removed_szqr = result_szqr.y[2, :]
quarantined_susceptible_szqr = result_szqr.y[3, :]
quarantined_zombie_szqr = result_szqr.y[4, :]

plt.plot(time_szqr, susceptible_szqr, label='susceptible', color='limegreen')
plt.plot(time_szqr, zombie_szqr, label='zombie', color='firebrick')
plt.plot(time_szqr, removed_szqr, label='removed', color='darkorchid')
plt.plot(time_szqr, quarantined_susceptible_szqr, label='quarantined susceptible', color='orange')
plt.plot(time_szqr, quarantined_zombie_szqr, label='quarantined zombie', color='purple')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('population (thousands)')
plt.title('Solution of the SZQR Model with Quarantine')
plt.show()
```
