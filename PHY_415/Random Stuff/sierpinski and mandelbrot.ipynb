```python
import matplotlib.pyplot as plt
import numpy as np

def plot_sierpinski(x, y, size, depth):
    if depth == 0:
        # Plot an equilateral triangle
        plt.fill(x, y, 'm')
    else:
        # Calculate midpoints of the sides
        x_mid = [(x[0] + x[1]) / 2, (x[1] + x[2]) / 2, (x[2] + x[0]) / 2]
        y_mid = [(y[0] + y[1]) / 2, (y[1] + y[2]) / 2, (y[2] + y[0]) / 2]

        # Recursively plot smaller triangles
        plot_sierpinski([x[0], x_mid[0], x_mid[2]], [y[0], y_mid[0], y_mid[2]], size / 2, depth - 1)
        plot_sierpinski([x_mid[0], x[1], x_mid[1]], [y_mid[0], y[1], y_mid[1]], size / 2, depth - 1)
        plot_sierpinski([x_mid[2], x_mid[1], x[2]], [y_mid[2], y_mid[1], y[2]], size / 2, depth - 1)

def main():
    size = 400  # Size of the initial equilateral triangle
    depth = 4  # Recursion depth (adjust as needed)

    # Define the coordinates of the vertices of the equilateral triangle
    x = np.array([0, size, size / 2, 0])
    y = np.array([0, 0, size * np.sqrt(3) / 2, 0])

    # Create a plot
    plt.figure(figsize=(6, 6))
    plt.axis('equal')
    plt.axis('off')

    # Plot the Sierpinski triangle
    plot_sierpinski(x, y, size, depth)

    # Show the plot
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the size of the image
width, height = 800, 800
x_min, x_max = -2.0, 1.0
y_min, y_max = -1.5, 1.5

# Create a grid of complex numbers
x = np.linspace(x_min, x_max, width)
y = np.linspace(y_min, y_max, height)
X, Y = np.meshgrid(x, y)
c = X + 1j * Y

# Initialize arrays to store the Mandelbrot set
z = c.copy()
mandelbrot_set = np.zeros((width, height))

# Set the maximum number of iterations
max_iterations = 256

for i in range(max_iterations):
    z = z * z + c
    mask = np.abs(z) < 1000  # Escape radius
    mandelbrot_set += mask

# Apply the Viridis colormap
plt.figure(figsize=(8, 8))
plt.imshow(np.log(mandelbrot_set + 1), extent=(x_min, x_max, y_min, y_max), cmap='viridis')
plt.colorbar()
plt.title('Mandelbrot Set (Viridis Colormap)')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Set the size of the image
width, height = 800, 800
max_iterations = 500

# Function to generate Mandelbrot set at a specific zoom level
def generate_mandelbrot(x_min, x_max, y_min, y_max):
    x = np.linspace(x_min, x_max, width)
    y = np.linspace(y_min, y_max, height)
    X, Y = np.meshgrid(x, y)
    c = X + 1j * Y

    z = c.copy()
    mandelbrot_set = np.zeros((width, height))

    for i in range(max_iterations):
        z = z * z + c
        mask = np.abs(z) < 1000  # Escape radius
        mandelbrot_set += mask

    return np.log(mandelbrot_set + 1)

# Define the animation parameters
frames = 50  # Number of frames in the animation
zoom_factor = 0.95  # Zoom factor per frame

# Define the initial zoom level
x_min, x_max, y_min, y_max = -2.0, 1.0, -1.5, 1.5

# Create and save each frame as an image
for frame in range(frames):
    mandelbrot_image = generate_mandelbrot(x_min, x_max, y_min, y_max)
    mandelbrot_image = (mandelbrot_image / np.max(mandelbrot_image) * 255).astype(np.uint8)
    image = Image.fromarray(mandelbrot_image, 'L')
    image.save(f'mandelbrot_frame_{frame:03d}.png')

    # Zoom in for the next frame
    x_center = (x_min + x_max) / 2
    y_center = (y_min + y_max) / 2
    x_range = (x_max - x_min) * zoom_factor
    y_range = (y_max - y_min) * zoom_factor
    x_min = x_center - x_range / 2
    x_max = x_center + x_range / 2
    y_min = y_center - y_range / 2
    y_max = y_center + y_range / 2

# Combine the frames into a GIF
import glob
import os

frames_filenames = sorted(glob.glob('mandelbrot_frame_*.png'))
images = [Image.open(filename) for filename in frames_filenames]
output_gif = 'mandelbrot_zoom.gif'
images[0].save(output_gif, save_all=True, append_images=images[1:], duration=100, loop=0)

# Clean up temporary image files
for filename in frames_filenames:
    os.remove(filename)

print(f'GIF saved as {output_gif}')
```
