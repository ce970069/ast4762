The subject of our investigation in this project is a sawtooth wave signal, characterized by its continuous linear rise and abrupt drop. This signal can be represented as a periodic function that repeats its pattern at specific intervals. Our primary objective is to employ Fourier series analysis to gain an understanding of the wave. The Fourier series enables us to break down complex signals into their fundamental sinusoidal components, allowing us to delve into the signal's underlying frequency characteristics. To achieve this, we will begin by computing the Fourier coefficients for the sawtooth wave. These coefficients will be determined by performing integrals over one period of the signal, utilizing the complex form of the Fourier series. These integrals will be evaluated numerically. The central focus of this project is to assess how accurately the Fourier series approximates the original sawtooth wave as we vary the number of terms $N$ in the series.

We will define the approximate form of the signal using the complex form of the Fourier series:

$$V(t) = \sum_{-\infty}^{+\infty} c_n e^{i n \omega_0 t}$$

where $\omega_0 = \frac{2 \pi}{T_0}$ and $c_n$ is the Fourier coefficient given by:

$$c_n = \frac{1}{T_0} \int_{0}^{T_0} V(t) e^{-i n \omega_0 t} dt$$

We will use the real $V(t)$ to compute the complex $c_n$. Once we have found the $c_n$ we can reconstruct the signal using the series expansion.

$$V(t) \approx V_{approx}(t) = \sum_{-N}^{+N} c_n e^{i n \omega_0 t}$$

where $N$ is the number of terms we use in the series expansion.

# Analysis on `scipy.signal.sawtooth` 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import sawtooth
```

```python
# Define the parameters
V0 = 1.0  # Replace with the actual value of V0
T0 = 1.0  # Replace with the actual value of T0
N = 10
n_values = np.arange(-N, N+1) 

# Define the time values over one period [0, T0]
t_values = np.linspace(0, T0, num=1000)

# Initialize an array to store the c_n values
c_n_values = []

# Perform numerical integration for each n
for n in n_values:
    # Calculate the integrand
    integrand = V0 * np.exp(-1j * n * 2 * np.pi * t_values / T0)
    
    # Numerical integration using np.trapz
    integral = np.trapz(integrand, t_values)
    
    # Append the result to the c_n_values list
    c_n_values.append(integral / T0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import sawtooth

# Compute the Fourier coefficient
def compute_cn(v, n, T, f0):
    t = np.linspace(0, T, num=1000)  # Time values for integration
    integrand = v(t) * np.exp(-1j * 2 * np.pi * n * f0 * t)
    cn = 1/T * np.trapz(integrand, t)
    return cn

# Fourier series expansion using complex form
def complex_fourier_series_expansion(v, t, T, N):
    f0 = 1 / T
    series = np.zeros_like(t, dtype=complex)

    for n in range(-N, N + 1):
        cn = compute_cn(v, n, T, f0)
        series += cn * np.exp(1j * 2 * np.pi * n * f0 * t)

    return series.real

# Define the square wave signal
T = 2 * np.pi  # Period
def v(t):
    return sawtooth(t)

# Test your code
t = np.linspace(0, 4 * np.pi, num=1000)
N_values = [1, 3, 5, 7, 9, 11]  # Values of N for testing

# Create subplots for each value of N
fig, axs = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle('Approximation with Different N Values', fontsize=24)

for i, N in enumerate(N_values):
    row = i // 3
    col = i % 3

    approx_signal = complex_fourier_series_expansion(v, t, T, N)
    axs[row, col].plot(t, v(t), label='Original Signal', color='m')
    axs[row, col].plot(t, approx_signal, label=f'Approximation (N={N})', color='b')
    axs[row, col].set_title(f'Approximation with N={N}', fontsize =20)
    axs[row, col].grid()
    
    # Set larger font size for x and y labels
    axs[row, col].set_xlabel('Time', fontsize=20)
    axs[row, col].set_ylabel('Amplitude', fontsize=20)

for ax in axs.ravel():
    ax.legend()

# Adjust the layout to prevent overlapping titles
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Function to plot the magnitude of Fourier coefficients as a function of n*omega0
def plot_magnitude_of_coefficients(N, T):
    f0 = 1 / T
    n_values = np.arange(-N, N + 1)
    magnitudes = []

    for n in n_values:
        cn = compute_cn(v, n, T, f0)
        magnitudes.append(abs(cn))

    plt.figure()
    plt.stem(n_values * f0, magnitudes, basefmt=" ", use_line_collection=True)
    plt.title('Magnitude of Fourier Coefficients')
    plt.xlabel('n*omega0', fontsize=12)
    plt.ylabel('|c_n|', fontsize=12)
    plt.grid()
    plt.show()

# Call the function to plot the magnitude of Fourier coefficients
plot_magnitude_of_coefficients(10, T)

plt.show()
```

From the graphs of our analysis, several key observations can be made: Firstly, as we increase the number of terms $N$ in the Fourier series approximation, the square wave signal is better approximated, and the approximation approaches the original signal more closely. This highlights the fundamental principle of Fourier analysis that a signal can be accurately represented as a sum of sinusoidal components. Additionally, the magnitude of Fourier coefficients graph shows distinct peaks at certain frequencies, with the magnitude decreasing as we move away from these peaks. These peaks correspond to the significant frequency components of the square wave signal. Secondly, the oscillations near the abrupt changes in the square wave, known as the **Gibbs phenomenon**, becomes more obvious as $N$ increases. This phenomenon is a characteristic feature of Fourier series approximations near discontinuities.
