## Analytical Solution Part I:

To find the electric field at a distance $z$ above the center of a flat circular disk of radius $R$ with a uniform surface charge $\sigma$, we can use Gauss's law and symmetry arguments. 

To calculate the electric field, we will apply Gauss's law:

$$
\oint \mathbf{E} \cdot \mathrm{d}\mathbf{A} = \frac{Q_{\text{enc}}}{\varepsilon_0}
$$

Where:
- $\mathbf{E}$ is the electric field.
- $\mathrm{d}\mathbf{A}$ is an infinitesimal area vector on the Gaussian surface.
- $Q_{\text{enc}}$ is the charge enclosed by the Gaussian surface.
- $\varepsilon_0$ is the vacuum permittivity.

$$
\oint \mathbf{E} \cdot \mathrm{d}\mathbf{A} = E \oint \mathrm{d}A = E \cdot 2\pi r z
$$

The charge enclosed within the Gaussian surface is the surface charge density $\sigma$ times the area of the top face of the cylinder, which is $\pi R^2$:

$$
Q_{\text{enc}} = \sigma \pi R^2
$$

Now, we can rewrite Gauss's law as:

$$
E \cdot 2\pi r z = \frac{\sigma \pi R^2}{\varepsilon_0}
$$

Solving for the electric field $E$, we get:

$$
E = \frac{\sigma R^2}{2 \varepsilon_0 z}
$$

This is the expression for the electric field at a distance $z$ above the center of the circular disk. 

Now, let's analyze the limit cases:

1. **$R \rightarrow \infty$**: In this limit, as the radius of the disk becomes infinitely large, the electric field becomes zero. This is because as $R$ approaches infinity, the disk becomes like an infinite plane with uniform charge density, and the electric field above an infinite plane is zero.

2. **$z \gg R$**: When the distance $z$ is much larger than the radius $R$, the electric field can be approximated as:

$$
E \approx \frac{\sigma}{2 \varepsilon_0 z}
$$

This is similar to the electric field above an infinitely large, uniformly charged plane.

3. **$R \gg z$**: When the radius $R$ is much larger than the distance $z$, the electric field can be approximated as:

$$
E \approx \frac{\sigma R}{2 \varepsilon_0 z^2}
$$

In this case, the electric field decreases as $z$ becomes smaller relative to the large radius of the disk.

## Plotting the Electric Field

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import matplotlib.cm as cm
```

```python
def electric_field_disk(sigma, R, x_points, y_points):
    k = 8.99e9  # Nm^2/C^2, Coulomb's constant
    
    # Initialize electric field components to zero
    E_x = np.zeros_like(x_points)
    E_y = np.zeros_like(y_points)
    
    # Calculate electric field components due to the charged disk at each point on the grid
    for i in range(x_points.shape[0]):
        for j in range(y_points.shape[1]):
            x = x_points[i, j]
            y = y_points[i, j]
            
            # Calculate the distance from the point (x, y) to the center of the disk (0, 0)
            r = np.sqrt(x**2 + y**2)
            
            # Check if the point is inside the disk
            if r <= R:
                # Calculate the electric field components due to an infinitesimal charge element on the disk
                dE = k * sigma * (x / r**3), k * sigma * (y / r**3)
                
                # Add the contribution from this charge element to the total electric field
                E_x[i, j] = dE[0]
                E_y[i, j] = dE[1]
    
    return E_x, E_y
```

```python
def plot_electric_field(X, Y, E_x, E_y, stream_color='b'):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    
    # First Subplot - Quiver Plot
    axs[0].quiver(X, Y, E_x, E_y, scale=1e6, color='r')
    axs[0].set_title('Quiver Plot')
    axs[0].set_xlabel('x [m]')
    axs[0].set_ylabel('y [m]')
    axs[0].axhline(0, color='black', linewidth=0.5)
    axs[0].axvline(0, color='black', linewidth=0.5)
    axs[0].grid(color='gray', linestyle='--', linewidth=0.5)
    
    # Second Subplot - Stream Plot
    axs[1].streamplot(X, Y, E_x, E_y, color=stream_color, linewidth=1, density=2, arrowstyle='->', arrowsize=1.5)
    axs[1].set_title('Stream Plot')
    axs[1].set_xlabel('x [m]')
    axs[1].set_ylabel('y [m]')
    axs[1].axhline(0, color='black', linewidth=0.5)
    axs[1].axvline(0, color='black', linewidth=0.5)
    axs[1].grid(color='gray', linestyle='--', linewidth=0.5)
    
    # Third Subplot - Color Scaled Quiver Plot
    magnitude = np.sqrt(E_x**2 + E_y**2)
    colors = np.log(magnitude)
    strm = axs[2].streamplot(X, Y, E_x, E_y, color=colors, cmap = 'gist_rainbow') #seismic? cmap god BLESS AMERICA CAWWWWW
    axs[2].set_title('Color Scaled Stream Plot')
    axs[2].set_xlabel('x [m]')
    axs[2].set_ylabel('y [m]')
    axs[2].axhline(0, color='black', linewidth=0.5)
    axs[2].axvline(0, color='black', linewidth=0.5)
    axs[2].grid(color='gray', linestyle='--', linewidth=0.5)
    # Add a colorbar to the plot
    cbar = plt.colorbar(strm.lines)
    cbar.set_label(r'$\log{\left|\mathbf{E}\right|}$')

    
    plt.tight_layout()  # Adjust spacing between subplots
    plt.show()
```

```python
# Set Grid of Points
x = np.linspace(-1, 1, 100)  # Adjust the range and resolution as needed
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

# Surface charge density and radius of the disk
sigma = 1e-6  # C/m^2
R = 1  # Radius of the circular disk in meters

# Calculate electric field components
E_x, E_y = electric_field_disk(sigma, R, X, Y)

# Plot the electric field
plot_electric_field(X, Y, E_x, E_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def electric_field_disk_3d(sigma, R, x_points, y_points, z_points):
    k = 8.99e9  # Nm^2/C^2, Coulomb's constant
    
    # Initialize electric field components to zero
    E_x = np.zeros_like(x_points)
    E_y = np.zeros_like(y_points)
    E_z = np.zeros_like(z_points)
    
    # Calculate electric field components due to the charged disk at each point on the grid
    for i in range(x_points.shape[0]):
        for j in range(y_points.shape[1]):
            for k in range(z_points.shape[2]):
                x = x_points[i, j, k]
                y = y_points[i, j, k]
                z = z_points[i, j, k]
                
                # Calculate the distance from the point (x, y, z) to the center of the disk (0, 0, 0)
                r = np.sqrt(x**2 + y**2 + z**2)
                
                # Check if the point is inside the disk
                if r <= R:
                    # Calculate the electric field components due to an infinitesimal charge element on the disk
                    dE_x = k * sigma * (x / r**3)
                    dE_y = k * sigma * (y / r**3)
                    dE_z = k * sigma * (z / r**3)
                    
                    # Add the contribution from this charge element to the total electric field
                    E_x[i, j, k] = dE_x
                    E_y[i, j, k] = dE_y
                    E_z[i, j, k] = dE_z
    
    return E_x, E_y, E_z

# Define the function to visualize the electric field in 3D
def plot_electric_field_3d(X, Y, Z, E_x, E_y, E_z):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Plot electric field vectors using a stream plot
    density = 2
    ax.quiver(X, Y, Z, E_x, E_y, E_z, length=0.5, normalize=True, color='b')

    # Customize plot appearance
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Electric Field in 3D (Stream Plot)')
    
    plt.show()

# Define the 3D grid
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
z = np.linspace(-2, 2, 20)
X, Y, Z = np.meshgrid(x, y, z)

# Surface charge density and radius of the disk
sigma = 1e-6  # C/m^2
R = 1  # Radius of the circular disk in meters

# Calculate electric field components in 3D
E_x, E_y, E_z = electric_field_disk_3d(sigma, R, X, Y, Z)

# Visualize the electric field in 3D using a stream plot
plot_electric_field_3d(X, Y, Z, E_x, E_y, E_z)
```

## Analytical Solution Part II: Multipole Expansion

### General Solution:

The electric field at a point P due to a charge distribution can be expressed as a series of moments. The general formula for the electric field at point P, due to a charge distribution, is given by:

$$\vec{E}(\vec{r}) = \frac{1}{4\pi\epsilon_0} \int \frac{\rho(\vec{r'}) (\vec{r} - \vec{r'})}{|\vec{r} - \vec{r'}|^3} d\tau'$$

Where:
- $\vec{E}(\vec{r})$ is the electric field at point P.
- $\epsilon_0$ is the vacuum permittivity.
- $\rho(\vec{r'})$ is the charge density at a point $\vec{r'}$.
- $\vec{r}$ is the position vector of point P.
- $\vec{r'}$ is the position vector of a point in the charge distribution.
- $d\tau'$ is an element of volume in the charge distribution.

We will now use the **Multipole Expansion Method** to find the moments of the distribution.

### Multipole Expansion Method:

1. **Monopole Term** (Leading Term):
   
   $$V(\vec r) = \frac{1}{4\pi\varepsilon_0} \int \frac{\rho(\vec r') d\tau'}{r} $$
   $$V(\vec r) = \frac{1}{4\pi\varepsilon_0} \frac{1}{r} \int \rho(\vec r') d\tau'$$
   $$ \therefore V(\vec r) \approxeq \frac{Q_{tot}}{4\pi\varepsilon_0r}$$

2. **Dipole Term** (Secondary Term, Linear Asymmetries):

   $$V(\vec r) = \frac{1}{4\pi\varepsilon_0} \int \frac{\rho(\vec r') d\tau'}{r} \varepsilon \cos{\theta '}$$
   $$V(\vec r) = \frac{1}{4\pi\varepsilon_0} \int \frac{\rho(\vec r') d\tau'}{r} \frac{r'}{r} \cos{\theta '}$$
   $$V(\vec r) = \frac{1}{4\pi\varepsilon_0 r^2} \int {\rho(\vec r') d\tau'} r' \cos{\theta '}$$
   $$V(\vec r) = \frac{Q_{tot}}{4\pi\varepsilon_0 r^2} \int r \hat r \cos{\theta '}$$
   $$\therefore V(\vec r) \approxeq \frac{Q_{tot} \cos{\theta '}}{8\pi\varepsilon_0}$$


   
3. **Boundary Conditions**:
\
    All of the following are satisfied with the equations above:
    1) The electric potential is finite at all points in space.
    2) The electric field is continuous across the surface of the circular disk.
    3) The electric field goes to zero at infinity. 
    $$$$
4. **Particular Solution** (Final Answer):

    $$V(\vec r) \approxeq \frac{Q_{tot}}{4\pi\varepsilon_0r} + \frac{Q_{tot} \cos{\theta '}}{8\pi\varepsilon_0}$$    
    $$V(\vec r) \approxeq \frac{Q_{\text{tot}}}{4\pi\varepsilon_0} \left(\frac{1}{r} + \frac{\cos(\theta')}{2}\right)$$


## Limits of the Problem

**Limit 1: $r \rightarrow \infty$**

As $r$ becomes approaches infinity, the dominant term in the electric potential expression is $\frac{1}{r}$, which is the monopole term. This is consistent with the behavior of a point charge. The potential decreases with increasing distance from the disk, following an inverse square law.

**Limit 2: $\theta' \rightarrow 0$ and $\theta' \rightarrow \pi$**

When $\theta' = 0$, the term $\frac{\cos(\theta')}{2}$ becomes $\frac{1}{2}$, and when $\theta' = \pi$, $\cos(\theta')$ becomes -1. Therefore, at these angles, the dipole term is maximized and minimized, respectively. At $\theta' = 0$, the electric potential is larger in the direction perpendicular to the plane of the disk, and at $\theta' = \pi$, it's smaller in the same direction. This is indicative of the anisotropic behavior introduced by the dipole term.

**Limit 3: $\theta' \rightarrow \frac{\pi}{2}$**

At $\theta' = \frac{\pi}{2}$, the dipole term vanishes $(\cos(\frac{\pi}{2}) = 0)$. This means that along the direction perpendicular to the plane of the disk, there is no contribution from the dipole moment. The electric potential in this direction is solely determined by the monopole term.

## Plotting the Electric Potential

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
R = 1.0  # Radius of the circular disk
z = 1.0  # Distance above the center of the disk
sigma = 1.0  # Surface charge density

# Number of grid points and range
steps = 100
x_range = np.linspace(-2 * R, 2 * R, steps)
y_range = np.linspace(-2 * R, 2 * R, steps)

# Create a mesh grid
X, Y = np.meshgrid(x_range, y_range)

# Function to calculate electric potential
def electric_potential(x, y, z, sigma, R):
    r = np.sqrt(x**2 + y**2)
    if r <= R:  # Check if the point is inside the disk
        potential = (sigma / (2 * 8.85e-12))  # Inside the disk (constant potential)
    else:
        potential = (sigma / (4 * 8.85e-12 * R)) / r  # Outside the disk
    return potential

# Calculate the electric potential on the mesh grid
Z = np.zeros_like(X)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = electric_potential(X[i, j], Y[i, j], z, sigma, R)

# Create a contour plot (heat map)
contour = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Electric Potential Heat Map')

# Add a black line around the circumference of the disk
circle = plt.Circle((0, 0), R, color='black', fill=False, linewidth=2)
plt.gca().add_patch(circle)

# Adjust the plot limits to zoom out
plt.xlim(-2 * R, 2 * R)
plt.ylim(-2 * R, 2 * R)

plt.show()
```

Inside the disk, we assume the potential is constant due to the uniform charge distribution. However, outside the disk, the electric potential follows our above calculations (due to a uniformly charged disk).

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define parameters
Q_tot = 20.0
r = 0.5
epsilon_0 = 8.854187817e-12

# Create a grid of points in polar coordinates
r_values = np.linspace(0.001, 2.0, 100)  # Avoid r=0 to prevent division by zero
theta_prime_values = np.linspace(0, 2 * np.pi, 100)
R, Theta = np.meshgrid(r_values, theta_prime_values)

# Calculate the electric potential at each point
V = (Q_tot / (4 * np.pi * epsilon_0)) * (1 / R + np.cos(Theta) / 2)

# Plot the electric potential surface
surf = ax.plot_surface(R * np.cos(Theta), R * np.sin(Theta), V, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Electric Potential (V)')
ax.set_title('Electric Potential Surface Plot')

plt.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

plt.show()
```
