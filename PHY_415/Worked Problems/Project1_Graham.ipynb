#  Lotka–Volterra Predator–Prey model

### Part 1: Describe the Differential Equation

For this assignment, I will be analyzing the Lotka-Volterra equations, or the Lotka-Volterra predator-prey model. It is a pair of 1st-order, nonlinear, differential equations which can be used to describe the dynamics of biological systems with interactions between a predator and its prey. These two populations change through time as per the following equations: 
$$\begin{align*}
\frac{dx}{dt} &= \alpha x - \beta xy \\
\frac{dy}{dt} &= \delta xy - \gamma y
\end{align*}$$

where $x$ is the population density of the prey, and $y$ for the predator. $\frac{dx}{dt}$ and $\frac{dy}{dt}$ represent the instantaneous growth rates of the two population, and $t$ is time. $\alpha$ and $\beta$ are parameters which describe the maximum prey per captia growth rate, as well as the effect of the presence of predators. $\gamma$ and $\delta$ are parameters which describe the predator's per captia death rate, as well as the effect of the presence of prey on the predator's growth rate. Also, $\alpha$, $\beta$, $\gamma$, $\delta$, $\in$ $\mathbf{R}^{+}$.

### Part 2: Analytical Investigation

$$\frac{dx}{dt} = \alpha x - \beta xy \text{ , } \frac{dy}{dt} = \delta xy - \gamma y$$

We will start by taking the limits on the nonlinear aspects:

| $y \approx 0$        | $x \approx 0$             | $x \gg y$                | $y \gg x$                |
|----------------------|---------------------------|--------------------------|--------------------------|
| $\dot{x} = \alpha x$ | $\dot{y} = \gamma y$      | $\dot{x} \approx \alpha x$ | $\dot{y} \approx -\beta x y$ |
| $x(t) = x_0 e^{\alpha t}$ | $y(t) = y_0 e^{-\gamma t}$ | $\dot{y} \approx \delta x y$ | $\dot{y} \approx -\gamma y$ |

We get four cases which we will analyze one by one:

$x(t) = x_0 e^{\alpha t}$: This is when no predators exist and the population of the prey explosed exponentially with respect to the per captia growth rate.
\

$y(t) = y_0 e^{-\gamma t}$: This is when no prey exists and the population of the predators decreases exponentially with respect to the per capita growth rate.
\

$\dot{x} \approx \alpha x$ and $\dot{y} \approx \delta x y$: When there are far more prey than predators, then the interactions are limited. All that is left is a positive growth rate of the prey and a positive death rate of the predators.
\

$\dot{y} \approx -\beta x y$ and $\dot{y} \approx -\gamma y$: When there are far more predators than prey, then the interactions are again limited. We see a negative growth term for the prey, which makes sense since they have ultimately no chance of thriving, and a negative death rate for the predators, which again makes sense since they will have unlimited resources to their advantage.

We will now investigate the approximate behaviors:

Depending of the variables $\alpha, \beta, \delta, \text{ and }, \gamma$ there are a few long-term projections that could be made of the system:

1) Stable Oscillations: Cyclic predator-prey relationship
2) Species Exctinction: Under certain conditions either the prey or the predator population could go exctinct.
3) Coexistence: Both species co-exist at some non-zero population without oscillating.

We will now perform limit analysis:

Using the results from Part 1, we can see that $x(t) = x_0 e^{\alpha t}$ and $y(t) = y_0 e^{-\gamma t}$ describe that of species exctinction due to the exponential term. In the case of $\dot{x} \approx \alpha x$ and $\dot{y} \approx \delta x y$, the prey completely dominates the predators and the population explodes and vice versa for $\dot{y} \approx -\beta x y$ and $\dot{y} \approx -\gamma y$.

With the correct parameters, the cyclic growth and death of the prey will complement those of the predators and form a closed loop of some shape.

### Fixed Points:

$$ \begin{align*}
(0,0) &\text{ . . . saddle point} \\
\left(\frac{\gamma }{\delta },\frac{\alpha }{\beta }\right) &\text{ . . . center for closed points}
\end{align*}$$

## Plotting the Phase Space

```python
#importing the needed libraries or whatever they're called in python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def LVPhasePortrait(x, y, alpha, beta, delta, gamma):
    '''LVPhasePortrait returns the value of
    the change in the phase variables at a given location
    in phase space for the LV model'''
    
    xdot = alpha*x - beta*x*y
    ydot = delta*x*y - gamma*y
    return xdot, ydot

def ComputeLVPhase(X, Y, alpha, beta, delta, gamma):
    '''ComputeLVPhase returns the changes in 
    the phase variables across a grid of locations
    that are specified'''
    
    # Prep the arrays with zeros at the right size
    xdot, ydot = np.zeros(X.shape), np.zeros(Y.shape)
    
    # Set the limits of the loop based on how many points in the arrays we have
    Xlim, Ylim = X.shape
    
    # Calculate the changes at each location and add them to the arrays
    for i in range(Xlim):
        for j in range(Ylim):
            xloc = X[i, j]
            yloc = Y[i, j]
            xdot[i, j], ydot[i, j] = LVPhasePortrait(xloc, yloc, alpha, beta, delta, gamma)
    return xdot, ydot

def VP_eqn_for_solve_ivp(t, curr_vals, alpha, beta, delta, gamma):
    '''VP_eqn_for_solve_ivp solves our differential equations 
    with respect to the parameters and the inital values that 
    I set later on in the code'''
    x, v = curr_vals 
    xdot, vdot = LVPhasePortrait(x, v, alpha, beta, delta, gamma) #calling our first function 
    return xdot, vdot
```

```python
# Setting parameters and the phase space variables
alpha = 1 #prey growth rate
beta = 0.5 #prey death rate
delta = 0.5 #predator growth rate 
gamma = 0.5 #predator death rate

x = np.linspace(0, 30, 50)
y = np.linspace(0, 30, 50)
X, Y = np.meshgrid(x, y)

# Run the Lotka-Volterra calculations
xdot, ydot = ComputeLVPhase(X, Y, alpha, beta, delta, gamma)

# Initial conditions
x0 = 10 #initial prey
y0 = 2 #initial predator

# Plot the phase space plot
fig, ax = plt.subplots(figsize=(10, 8))

# Plot with Quiver
Q = plt.streamplot(X, Y, xdot, ydot, color='k', broken_streamlines = False)
plt.xlabel('$x$: (Population of Prey)', fontsize=16)
plt.ylabel('$y$: (Population of Predators)', fontsize=16)
plt.title('Lotka-Volterra Phase Space', fontsize=18)

# Plot initial conditions
plt.plot(x0, y0, 'bo', markersize=9, label='Initial Condition')
plt.plot(0, 0, 'mo', markersize=9, label='F.P. - Saddle')
plt.plot((gamma/delta), (alpha/beta), 'ro', markersize=9, label='F.P. - Center for Closed Points')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

As we determined previously from the analytical investigation, with the correct parameters, the cyclic growth and death of the prey will complement those of the predators and form a closed loop of some shape. In this case, we have some sort of ovoid shape.

## Numerical Integration for Trajectories

```python
# Setting parameters and the phase space variables
alpha = 1  # prey growth rate
beta = 0.5  # prey death rate
delta = 0.5  # predator growth rate
gamma = 0.5  # predator death rate

# Numerical Integration
tmax = 50
dt = 0.05
tspan = (0, tmax)
t = np.arange(0, tmax, dt)

# Define initial conditions for multiple trajectories
# These are all random
initial_conditions = [
    [10, 2],
    [5, 8],
    [15, 3]
]

# Create a list to store the trajectories and labels for the legend
trajectories = []
labels = []

# Solve the equations for each initial condition
for i, initial_condition in enumerate(initial_conditions):
    solved = solve_ivp(
        VP_eqn_for_solve_ivp, tspan, initial_condition, t_eval=t,
        args=(alpha, beta, delta, gamma,), method="RK45"
    )
    trajectories.append(solved)
    labels.append(f'Trajectory {i + 1}')

# Plotting stuff
N = 30
x = np.linspace(0, 30, N)
y = np.linspace(0, 30, N)
X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots(figsize=(10, 8))
xdot, vdot = ComputeLVPhase(X, Y, alpha, beta, delta, gamma)
Q = plt.streamplot(X, Y, xdot, vdot, color='k', broken_streamlines=False)

# Plot each trajectory with a label
for i, solved in enumerate(trajectories):
    plt.plot(solved.y[0], solved.y[1], lw=3, label=labels[i])

# Add a legend outside the plot to the right
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.grid()
plt.xlabel('$x$: (Population of Prey)', fontsize=16)
plt.ylabel('$y$: (Population of Predators)', fontsize=16)
plt.title('Lotka-Volterra Phase Space with Trajectories', fontsize=18)
plt.show()
```
