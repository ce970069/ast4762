### Matrix Representation

In this part of the project, we will set up and develop the matrix representation for the given three-pendulum system. This matrix representation is essential for solving the system's equations of motion and finding the normal modes.

We start by defining the generalized coordinates for our system. In our case, we have three pendula, and the angles of these pendula are our generalized coordinates. We can represent these angles as a vector:

$$\boldsymbol{\theta} = \begin{bmatrix} \theta_1 \\ \theta_2 \\ \theta_3 \end{bmatrix}$$

Where:
- $\theta_1$ is the angle of the first pendulum.
- $\theta_2$ is the angle of the second pendulum.
- $\theta_3$ is the angle of the third pendulum.

The equations of motion for our system have been derived and are given as:

$$
\begin{align*}
\ddot{\theta}_1 &= -\frac{2k}{m}\theta_1 + \frac{k}{m}\theta_2 \\
\ddot{\theta}_2 &= -\frac{2k}{m}\theta_2 + \frac{k}{m}\theta_1 \\
\ddot{\theta}_3 &= -\frac{2k}{m}\theta_3
\end{align*}
$$

To set up the matrix representation, we express these equations in a matrix form $\ddot{\boldsymbol{\theta}} = \mathbf{A}\boldsymbol{\theta}$, where $\mathbf{A}$ is the coefficient matrix. We can write $\mathbf{A}$ as follows:

$$\mathbf{A} = \frac{b}{g} \begin{bmatrix}
1 - \varepsilon & -\varepsilon & -\varepsilon \\
-\varepsilon & 1 - \varepsilon & -\varepsilon \\
-\varepsilon & -\varepsilon & 1 - 2\varepsilon
\end{bmatrix}$$

Here, $\frac{b}{g}$ represents the mass and gravitational parameters, and $\varepsilon$ represents the coupling coefficient involving all three pendula.

The matrix representation of our system's equations of motion is:

$$\ddot{\boldsymbol{\theta}} = \frac{b}{g} \begin{bmatrix}
1 - \varepsilon & -\varepsilon & -\varepsilon \\
-\varepsilon & 1 - \varepsilon & -\varepsilon \\
-\varepsilon & -\varepsilon & 1 - 2\varepsilon
\end{bmatrix} \boldsymbol{\theta}$$

We have successfully set up the matrix representation, which is crucial for finding the normal modes and eigenfrequencies of the system.

### Finding the Normal Modes:

In this part of the project, we will use a matrix analysis approach to find the normal modes of the three-pendulum system.

Starting from the equations of motion for our system, which are given by:

$$
\begin{align*}
\ddot{\theta}_1 &= -\frac{2k}{m}\theta_1 + \frac{k}{m}\theta_2 \\
\ddot{\theta}_2 &= -\frac{2k}{m}\theta_2 + \frac{k}{m}\theta_1 \\
\ddot{\theta}_3 &= -\frac{2k}{m}\theta_3
\end{align*}
$$

We can rewrite them as a set of matrix equations, where $\boldsymbol{\theta} = \begin{bmatrix}\theta_1\\\theta_2\\\theta_3\end{bmatrix}$ represents the vector of angles:

$$\ddot{\boldsymbol{\theta}} = \begin{bmatrix}-\frac{2k}{m} & \frac{k}{m} & 0 \\ \frac{k}{m} & -\frac{2k}{m} & 0 \\ 0 & 0 & -\frac{2k}{m}\end{bmatrix}\boldsymbol{\theta}$$

Here, the coefficient matrix $\mathbf{A}$ is given by $\mathbf{A} = \begin{bmatrix}-\frac{2k}{m} & \frac{k}{m} & 0 \\ \frac{k}{m} & -\frac{2k}{m} & 0 \\ 0 & 0 & -\frac{2k}{m}\end{bmatrix}$.

To find the normal modes of the system, we need to calculate the eigenvalues and eigenvectors of the coefficient matrix $\mathbf{A}$. The eigenvalues represent the frequencies of the normal modes, while the eigenvectors indicate the relative amplitudes of the pendula in each normal mode.

To find the eigenvalues, we solve the characteristic equation:

$$\det(\mathbf{A} - \omega^2\mathbf{I}) = 0$$

Where $\mathbf{I}$ is the identity matrix. Solving this equation for $\omega^2$ gives us the eigenfrequencies, which are the squared angular frequencies of the normal modes.

The normal modes of the system are specific patterns of motion in which the three pendula oscillate at certain frequencies and with particular relative amplitudes. These modes represent the system's natural vibrations. For our system, there are three eigenfrequencies $\omega_1$, $\omega_2$, and $\omega_3$, each corresponding to a different normal mode. The eigenvectors associated with these eigenfrequencies provide information about the relative amplitudes of the pendula in each mode.

- Mode 1 ($\omega_1$): This mode corresponds to the first eigenfrequency $\omega_1$. The eigenvector associated with this mode will give us the relative amplitudes of the angles $\theta_1$, $\theta_2$, and $\theta_3$ in this particular mode.
<br>
<br>
- Mode 2 ($\omega_2$): This mode corresponds to the second eigenfrequency $\omega_2$. The eigenvector associated with this mode will provide us the information about how the angles $\theta_1$, $\theta_2$, and $\theta_3$ oscillate in this mode.
<br>
<br>
- Mode 3 ($\omega_3$): This mode corresponds to the third eigenfrequency $\omega_3$. The eigenvector for this mode will describe the relative amplitudes and motion of the pendula in this particular mode.

I will now solve for the eigenvalues and eigenvectors of the coefficient matrix $\mathbf{A}$ below:

```python
import numpy as np
from scipy.linalg import eigh

# Define parameters
m = 1.0  # Mass
k = 1.0  # Spring constant

# Define the coefficient matrix
A = np.array([[-2 * k / m, k / m, 0],
              [k / m, -2 * k / m, 0],
              [0, 0, -2 * k / m]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = eigh(A)

# Print the results
print("Eigenvalues (squared frequencies of normal modes):")
print(eigenvalues)
print("\nEigenvectors (relative amplitudes of the masses for each mode):")
print(eigenvectors)
```

## Plotting the Motion of the System for a Given Set of Initial Conditions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters for your oscillatory system
m = 1.0  # Mass
k = 1.0  # Spring constant
e = 0.1  # Coupling coefficient of the bar

# Define the system of differential equations
def system(t, theta):
    theta1, theta2, theta3 = theta
    # Define the differential equations based on the system's equations of motion
    dtheta1dt = -2 * k / m * theta1 + k / m * theta2
    dtheta2dt = -2 * k / m * theta2 + k / m * theta1
    dtheta3dt = -2 * k / m * theta3
    return [dtheta1dt, dtheta2dt, dtheta3dt]

# Initial conditions
initial_conditions = [0.5, 0.1, 0.1]

# Time span for simulation
t_span = (0, 10)  

# Solve the system for the chosen initial conditions
sol = solve_ivp(system, t_span, initial_conditions, t_eval=np.linspace(*t_span, 1000))

# Plot the motion of the system
plt.figure()
plt.plot(sol.t, sol.y[0], label="Theta 1")
plt.plot(sol.t, sol.y[1], label="Theta 2")
plt.plot(sol.t, sol.y[2], label="Theta 3")
plt.xlabel("Time")
plt.ylabel("Displacement")
plt.legend()
plt.title("Motion of the System")
plt.grid(True)
plt.show()
```
