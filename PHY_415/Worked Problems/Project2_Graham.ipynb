## System Description:

### Problem 3.5 (Griffith's): An infinitely long rectangular metal pipe (sides $a$ and $b$) is grounded, but one end, at $x=0$, is maintained at a specified potential $V_{0}(y,z)$, as indicated in the figure below. Find the potential inside the pipe.

## Analytical Investigation:

We will begin with solving Laplace's equation:
$$\frac{\partial ^2 V}{\partial x^2} + \frac{\partial ^2 V}{\partial y^2} + \frac{\partial ^2 V}{\partial z^2} = 0 $$
which is subject to the following boundary conditions:
$$
\begin{align*}
    (i) & \quad V = 0 \text{ when } y = 0 \\
    (ii) & \quad V = 0 \text{ when } y = a \\
    (iii) & \quad V = 0 \text{ when } z = 0 \\
    (iv) & \quad V = 0 \text{ when } z = b \\
    (v) & \quad V \rightarrow 0 \text{ as } x \rightarrow \infty \\
    (vi) & \quad V = V_0(y, z) \text{ when } x = 0 \\
\end{align*}
$$

<br>
<br>
Our first step will be to look for solutions in the form of products:
$$V(x,y,z) = X(x)Y(y)Z(z)$$
Plugging this into our first expression, we get the following:
$$\frac{1}{X}\frac{\partial ^2 V}{\partial x^2} + \frac{1}{Y}\frac{\partial ^2 V}{\partial y^2} + \frac{1}{Z}\frac{\partial ^2 V}{\partial z^2} = 0$$
Where it follows that
$$\frac{1}{X}\frac{\partial ^2 V}{\partial x^2} = C_1,  \frac{1}{Y}\frac{\partial ^2 V}{\partial y^2} =C_2,  \frac{1}{Z}\frac{\partial ^2 V}{\partial z^2} = C_3$$
such that
$$C_1 + C_2 + C_3 = 0.$$

<br>
<br>
Previous examples in the past, such as the infinite potential well, suggest that $C_1$ must be positive, while $C_2$ and $C_3$ must be negative. Setting $C_2 = -k^2$ and $C_3 = -l^2$, we have that $C_1 = k^2 + l^2$, and therefore:
$$\frac{d^2X}{dx^2} = (k^2+l^2)X, \frac{d^2Y}{dy^2} = -k^2Y, \frac{d^2Z}{dz^2} = -l^2Z.$$
<br>
Whose solutions are the following:
<br>
$$X(x)= Ae^{\sqrt{k^2+l^2}}+Be^{-\sqrt{k^2+l^2}}$$
$$Y(y) = C\sin {ky} + D\cos {ky}$$
$$Z(z) = E \sin {lz} + F \cos {lz}$$
<br>
From our boundary conditions we have that:
$$
\begin{align*}
    (v)& \text{ implies that } A=0 \\
    (i)& \text{ implies that } D = 0 \\
    (iii)& \text{ implies that } F = 0 \\
    (ii)& \text{ implies that } k = n \pi /a  \text{ where } n \in \mathbb R^+\\ 
    (iv)& \text{ implies that } l = m \pi / b \text{ where }  m \in \mathbb R^+
\end{align*}
$$
<br>
Combining the remaining constants, we are left with the following expression:
$$V(x,y,z) = C e ^{-\pi \sqrt{(n/a)^2+(m/b)^2x}}\sin (n\pi y/a)\sin (m\pi z/b) $$
<br>
Notice that this meets all the boundary conditions expect that of $(vi)$, that $V = V_0(y, z) \text{ when } x = 0$. It fails to meet this condition due to the two unspecified integers, $n$ and $m$. Therefore, the most general linear combination is a double sum in the following form:
$$V(x,y,z)=\sum_{n=1}^{\infty}\sum_{m=1}^{\infty}C_{n,m} e ^{-\pi \sqrt{(n/a)^2+(m/b)^2x}}\sin (n\pi y/a)\sin (m\pi z/b)$$
<br>
Where we aim to fit the remaining boundary condition:
$$V(0,y,z)=\sum_{n=1}^{\infty}\sum_{m=1}^{\infty}C_{n,m} \sin (n\pi y/a)\sin (m\pi z/b) = V_0(y,z)$$
<br>
<br>

In order to to do this, we must have an appropriate choice of the coefficients $C_{n,m}$. To determine these constants, we must multiply by $\sin (n'\pi y /a)\sin (m'\pi z/b)$, where $m', n' \in \mathbb R^+$, and then integrate accordingly:
$$\sum_{n=1}^{\infty}\sum_{m=1}^{\infty}C_{n,m}\int_{0}^{a} \sin(n\pi y/a)\sin(n'\pi y/a) \int_{0}^{b}\sin (m\pi z/b)\sin (m'\pi z/b)dz$$
$$=\int_{0}^{a}\int_{0}^{b}V_0(y,z)\sin(n'\pi y/a) \ \sin (m'\pi z/b) \ dy \ dz$$
$$=\frac{4}{ab}\int_{0}^{a}\int_{0}^{b}V_0(y,z)\sin(n\pi y/a) \ \sin (m\pi z/b) \ dy \ dz$$
$$C_{n,m}=\frac{4V_0}{ab} \int_{0}^{a}\sin(n\pi y/a) \ dy \int_{0}^{b}\sin (m\pi z/b) \ dz$$


$$
=\begin{cases}
0 & \text{if n or m is even} \\
\frac{16V_0}{\pi^2 nm} & \text{if n and m are odd}
\end{cases}
$$

Therefore, our final solution is the following expression:
$$
\boxed{V(x,y,z) = \frac{16V_0}{\pi^2}\sum_{n,m = 1,3,5 ...}^{\infty} \frac{1}{nm}e ^{-\pi \sqrt{(n/a)^2+(m/b)^2x}}\sin (n\pi y/a)\sin (m\pi z/b)}
$$

<br>
<br>

Notice that this solution gives us an **infinite potential**, which was obvious by the problem description. For the sake of the rest of the project, we will establish a boundary in the x-dimension that will allow us to take limits and use the relaxation method later on.
<br>
<br>
Instead of having our pipe extent infinitely in the x-direction, we will have it end when $x=a$, creating a rectangular prism with the following boundary conditions:
$$
\begin{align*}
    (i) & \quad V = 0 \text{ when } x = 0 \\
    (ii) & \quad V = 0 \text{ when } x = a \\
    (iii) & \quad V = 0 \text{ when } y = 0 \\
    (iv) & \quad V = 0 \text{ when } y = a \\
    (v) & \quad V = 0 \text{ when } z = 0 \\
    (vi) & \quad V = V_0 \text{ when } z=b\\
\end{align*}
$$
<br>
Following our solution process above, we have the following solutions:
$$X(x)= Ae^{\sqrt{k^2+l^2}}+Be^{-\sqrt{k^2+l^2}}$$
$$Y(y) = C\sin {ky} + D\cos {ky}$$
$$Z(z) = E \sin {lz} + F \cos {lz}$$
<br>
From our boundary conditions we have that:
$$
\begin{align*}
    (i)& \text{ implies that } D = 0 \\
    (iii)& \text{ implies that } F = 0 \\
    (ii)& \text{ implies that } k = n \pi /a  \text{ where } n \in \mathbb R^+\\ 
    (iv)& \text{ implies that } l = m \pi / b \text{ where }  m \in \mathbb R^+ \\
    (v)& \text{ implies that } A+B=0 \\
\end{align*}
$$
And therefore,
$$X(x) = 2E\sinh (\pi \sqrt{n^2+m^2}x/a)$$

<br>
<br>
Now combining everything with our constants,
$$V(x,y,z)=\sum_{n=1}^{\infty}\sum_{m=1}^{\infty}C_{n,m} \sin (n\pi y/a)\sin (m\pi z/b) \sinh (\pi \sqrt{n^2+m^2}x/a)$$
<br>
Now we need to evaluate the consants $C_{n,m}$ by imposing boundary condition $(vi)$ where $V = V_0 \text{ when } z=b$:
$$V_0 = \sum \sum \left [ C_{n,m}\sinh(\pi \sqrt{n^2+m^2}) \right] \sin (n\pi y/a) \ \sin (m\pi z/b)$$
\begin{align*}
C_{n,m}\sinh (\pi \sqrt{n^2+m^2}) &= \left (\frac{2}{a} \right )^2 V_0 \int_{0}^{a} \int_{0}^{b} \sin (n\pi y/a) \ \sin (m\pi z/b) \ dx \ dy \\
&=\begin{cases}
0 & \text{if n or m is even} \\
\frac{16V_0}{\pi^2 nm} & \text{if n and m are odd}
\end{cases}
\end{align*}

<br>
Therefore:
$$
\boxed{ V(x,y,z) = \frac{16V_0}{\pi^2}\sum_{n=1,3,5...}^{\infty}\sum_{m=1,3,5...}^{\infty} \frac{1}{nm} \sin (n\pi y/a) \ \sin (m\pi z/b) \frac{\sinh \pi \sqrt{n^2+m^2} \ x/a}{\sinh \pi \sqrt{n^2+m^2}}}
$$
<br>
But this is also an infinite solution since we did not formally establish the length of the sides of the rectangular pipe, so we will begin this process now.

<br>
Let $n \equiv 2i +1$, $m \equiv 2j +1$, and note that $\sinh(2u) = 2 \sinh(u) \cosh(u)$. The double sum is then:
$$S = \frac{1}{2}\sum_{i=0}^{\infty}\sum_{j=0}^{\infty} \frac{(-1)^{i+j}}{(2i+1)(2j+1)} \ \text{sech} \left[ \frac{\pi \sqrt{(2i+1^2)+(2j+1)^2}}{2}\right]$$
<br>
Now setting the upper limits at $i = 3$, $j=3$, we get that $S = \frac{\pi^2}{96}$ or that:
<br>
<br>
$$
\boxed{ V \left( \frac{a}{2}, \frac{a}{2}, \frac{a}{2} \right) = \frac{V_0}{6}}
$$
<br>
Note that we set our newly imposed $x$ boundary $b = \frac{a}{2}$, so that our new shape is equal to that of a square. We will now start the limit analysis and relaxation process.

## Limit Analysis:

1. **Far from the boundary limits**:

   In this limit, we want to see how the potential behaves as you move to a large distance from the pipe. Recall our expression for $V(x, y, z)$:

   $$V(x, y, z) = \frac{16V_0}{\pi^2}\sum_{n,m = 1,3,5...}^{\infty} \frac{1}{nm}e ^{-\pi \sqrt{(n/a)^2+(m/b)^2x}}\sin \left(\frac{n\pi y}{a}\right)\sin \left(\frac{m\pi z}{b}\right)$$

   To find the limit as $x$ approaches infinity, now consider the exponential term:

   $$\lim_{x\to\infty} e^{-\pi \sqrt{(n/a)^2+(m/b)^2x}} = 0$$

   Therefore, the potential indeed approaches zero as you move far away from the pipe, as expected.
<br>

2. **Limit as the pipe size approaches zero**:

   For this limit, we must consider the behavior as both $a$ and $b$ approach zero. In this case, the potential simplifies to a point source:

   $$\lim_{a,b\to 0} V(x, y, z) = \frac{kQ}{r}$$

   This limit shows that when the dimensions of the pipe become very small and behaves like the electric potential from a point charge.
<br>

3. **Limit as the pipe size approaches infinity**:

   As $a$ and $b$ approach infinity, the potential should approach a constant, which is the potential applied at the end of the pipe:

   $$\lim_{a,b\to\infty} V(x, y, z) = V_0$$

   This demonstrates that our solution is consistent with the expected behavior when the pipe becomes infinitely large.

## Computational Investigation: Relaxation Method (2D and 3D)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def initialize_grid_2d(size):
    '''Initialize the grid of potential values'''
    return np.zeros((size, size))

def set_boundary(phi, edge, value, V_top=None):
    '''Set the boundary conditions for the grid'''
    '''phi is the whole grid of potential values that you initialized
    edge is the edge of the grid that you want to set (top, bottom, left, right)
    value is the value you want to set the edge to (single number or array)
    V_top is a specific value for the top edge (only used if edge is 'top')'''
    
    if edge == 'top':
        if V_top is not None:
            phi[-1, :] = V_top
        else:
            phi[-1, :] = value
    elif edge == 'bottom':
        phi[0, :] = value
    elif edge == 'left':
        phi[:, 0] = value
    elif edge == 'right':
        phi[:, -1] = value
        
    return phi

def relax(phi, N, tolerance, max_iterations=10000, store_frequency=1000):
    '''Relax the grid to solve Laplace's equation'''
    '''phi is the grid of potential values that you initialized
    after you set the boundary conditions. That is key!
    N is the size of the grid (assumed square
    tolerance is the stopping criterion - you control this; be careful
    max_iterations is the maximum number of iterations to run (stops loop if not converged)
    store_frequency is how often you want to store the phi values (for plotting)'''
    
    iterations = 0
    delta = 1.0
    stored_phi = []
    stored_deltas = []

    while delta > tolerance and iterations < max_iterations:
        phi_old = phi.copy()

        for i in range(1, N):
            for j in range(1, N):
                phi[i, j] = 0.25 * (phi_old[i+1, j] + phi_old[i-1, j] + phi_old[i, j+1] + phi_old[i, j-1])

        delta = np.max(np.abs(phi - phi_old))
        iterations += 1

        if iterations % store_frequency == 0:
            stored_phi.append(phi.copy())
            stored_deltas.append(delta)
            print(f"Iteration: {iterations}, Delta: {delta}")

    return stored_phi, stored_deltas, iterations
```

```python
# Example Usage
L = 1.0
h = 0.01
tolerance = 1e-5
N = int(L/h)

phi = initialize_grid_2d(N+1)

x = np.linspace(0, L, N+1)
V_0 = 6  
set_boundary(phi, 'top', value=V_0 / 6)
set_boundary(phi, 'bottom', 0)
set_boundary(phi, 'left', 5)
set_boundary(phi, 'right', 5)

# Relaxation
stored_phi, stored_deltas, iterations = relax(phi, N, tolerance)
print(f"Converged in {iterations} iterations.")

store_frequency = 1000

# 3D Visualization
fig = plt.figure(figsize=(12, 5))

# Create a 2D subplot
ax1 = fig.add_subplot(121)
ax1.imshow(phi, cmap='viridis', extent=[0, L, 0, L])
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('2D Relaxation')

# Create a 3D subplot
ax2 = fig.add_subplot(122, projection='3d')
x = np.linspace(0, L, N+1)
y = np.linspace(0, L, N+1)
X, Y = np.meshgrid(x, y)

for i in range(len(stored_phi)):
    ax2.plot_surface(X, Y, stored_phi[i], cmap='viridis', label=f"Iteration {i * store_frequency}")

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Potential')
ax2.set_title('3D Relaxation')

plt.tight_layout()
plt.show()
```

## Error Analysis

To compare the numerical solution to the analytical solution, I calculated the Root Mean Square Error (RMSE) in the cell below:

```python
V_analytical = V_0 / 6  # The analytical potential value at (1, 1, 1)

# Calculate absolute differences
absolute_differences = np.abs(phi - V_analytical)

# Calculate RMSE
mse = np.mean(absolute_differences**2)
rmse = np.sqrt(mse)
print(f"RMSE: {rmse}")
```

Therefore, the average difference between analytical values and the numerical values is 2.111.
