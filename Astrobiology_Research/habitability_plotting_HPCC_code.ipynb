## Desulfovibrio Plots

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def plot_ovoid(ax, center, radii, color, alpha=0.2):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    u, v = np.meshgrid(u, v)
    x = radii[0] * np.cos(u) * np.sin(v) + center[0]
    y = radii[1] * np.sin(u) * np.sin(v) + center[1]
    z = radii[2] * np.cos(v) + center[2]

    ax.plot_surface(x, y, z, color=color, alpha=alpha, rstride=1, cstride=1)

def get_species_data():
    species_data = {
        'Desulfovibrio JC010': { 
            'temperature': (4, 37),
            'pH': (4.0, 9.0),
            'salinity': (1.0, 5.5),
            'optimum': (28, 7.0, 4.0),
        },
        'Desulfovibrio JC022': {
            'temperature': (10, 37),
            'pH': (4.0, 8.0),
            'salinity': (2.5, 6.0),
            'optimum': (28, 7.0, 3.0),
        },
        'Desulfovibrio salexigans': {
            'temperature': (20, 40),
            'pH': (6.7, 8.2),
            'salinity': (0.5, 12),
            'optimum': (33, 7.8, 3.0),
        },
        'Desulfovibrio zosterae': {
            'temperature': (5, 34.5),
            'pH': (5.5, 7.5),
            'salinity': (0.0, 3.5),
            'optimum': (33.5, 7.1, 1.2),
        },
        'Desulfovibrio salinus': {
            'temperature': (15, 45),
            'pH': (6.0, 8.5),
            'salinity': (1, 12),
            'optimum': (40, 6.7, 3.0),
        },
        'Desulfovibrio bastinii': {
            'temperature': (20, 50),
            'pH': (5.2, 7.4),
            'salinity': (1, 12),
            'optimum': (37, 6, 4),
        },
        'Desulfovibrio ferrireducens': {
            'temperature': (-2, 30),
            'pH': (6.3, 7.5),
            'salinity': (0.7, 4),
            'optimum': (23, 7.3, 1.75),
        },
        'Desulfovibrio frigidus': {
            'temperature': (-2, 25),
            'pH': (6.9, 7.5),
            'salinity': (2, 3.5),
            'optimum': (21.5, 7.1, 2.5),
         },
        'Desulfovibrio hydrothermalis': {
            'temperature': (20, 40),
            'pH': (6.7, 8.2),
            'salinity': (2, 4),
            'optimum': (35, 7.8, 2.5),
        }
    }
    return species_data
```

```python
def plot_species_ovoids(species_data, view_angles):
    num_species = len(species_data)

    # Create a colormap with a color for each species
    colormap = plt.cm.get_cmap('tab10', num_species)

    for view_angle in view_angles:
        # Creating the 3D plot with a larger size
        fig = plt.figure(figsize=(14, 10))
        ax = fig.add_subplot(111, projection='3d')

        handles = []  # To store legend handles
        labels = []   # To store legend labels

        for idx, (species_name, data) in enumerate(species_data.items()):
            center = data['optimum']
            radii = (data['temperature'][1] - data['temperature'][0],
                     data['pH'][1] - data['pH'][0],
                     data['salinity'][1] - data['salinity'][0])
            radii = (radii[0] / 2.0, radii[1] / 2.0, radii[2] / 2.0)

            color = colormap(idx / num_species)

            plot_ovoid(ax, center, radii, color=color, alpha=0.2)
            label_offset = 0.5
            ax.text(center[0], center[1], center[2] + label_offset, species_name, fontsize=10, color=color, fontdict={'style': 'italic'})

            # Store handles and labels for the legend
            handles.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10))
            labels.append(species_name)

        ax.set_xlabel('Temperature (deg C)')
        ax.set_ylabel('pH')
        ax.set_zlabel('Salinity (NaCl %)')

        ax.view_init(elev=view_angle[0], azim=view_angle[1])

        # Create a legend on the middle left side, with additional left offset
        ax.legend(handles, labels, loc='center left', bbox_to_anchor=(0.1, 0.8), bbox_transform=plt.gcf().transFigure, prop={'style': 'italic'})

        plt.show()

species_data = get_species_data()

# List of different view angles (elev, azim)
view_angles = [
    (0, 270),  # Example view angle 1
    (20, 45),  # Example view angle 2
]

plot_species_ovoids(species_data, view_angles)
```
