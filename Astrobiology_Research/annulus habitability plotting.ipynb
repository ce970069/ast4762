```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to create a sphere
def plot_sphere(ax, radius, color, alpha):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = radius * np.outer(np.cos(u), np.sin(v))
    y = radius * np.outer(np.sin(u), np.sin(v))
    z = radius * np.outer(np.ones(np.size(u)), np.cos(v))
    ax.plot_surface(x, y, z, color=color, alpha=alpha, linewidth=0)

# Function to create an annulus
def plot_annulus(ax, inner_radius, outer_radius, color, alpha):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = outer_radius * np.outer(np.cos(u), np.sin(v))
    y = outer_radius * np.outer(np.sin(u), np.sin(v))
    z = outer_radius * np.outer(np.ones(np.size(u)), np.cos(v))
    ax.plot_surface(x, y, z, color=color, alpha=alpha, linewidth=0)

    x = inner_radius * np.outer(np.cos(u), np.sin(v))
    y = inner_radius * np.outer(np.sin(u), np.sin(v))
    z = inner_radius * np.outer(np.ones(np.size(u)), np.cos(v))
    ax.plot_surface(x, y, z, color=color, alpha=alpha, linewidth=0)

# Create a 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the semi-transparent sphere representing the planet
plot_sphere(ax, radius=1, color='grey', alpha=0.3)

# Plot the annuluses representing the habitability space of microbe species
plot_annulus(ax, inner_radius=0.3, outer_radius=0.5, color='green', alpha=0.3)
plot_annulus(ax, inner_radius=0.8, outer_radius=0.9, color='orange', alpha=0.3)

# Set plot limits and labels
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
