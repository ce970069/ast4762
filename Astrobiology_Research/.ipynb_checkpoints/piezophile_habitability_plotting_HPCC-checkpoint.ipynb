```python
def plot_species_ovoids(species_data, view_angles):
    num_species = len(species_data)

    # Create a colormap with a color for each species
    colormap = plt.cm.get_cmap('tab10', num_species)

    for view_angle in view_angles:
        # Creating the 3D plot with a larger size
        fig = plt.figure(figsize=(14, 10))
        ax = fig.add_subplot(111, projection='3d')

        handles = []  # To store legend handles
        labels = []   # To store legend labels

        for idx, (species_name, data) in enumerate(species_data.items()):
            center = data['optimum']
            radii = (data['temperature'][1] - data['temperature'][0],
                     data['pressure'][1] - data['pressure'][0],
                     data['salinity'][1] - data['salinity'][0])
            radii = (radii[0] / 2.0, radii[1] / 2.0, radii[2] / 2.0)

            color = colormap(idx / num_species)

            plot_ovoid(ax, center, radii, color=color, alpha=0.2)
            label_offset = 0.5
            ax.text(center[0], center[1], center[2] + label_offset, species_name, fontsize=10, color=color, fontdict={'style': 'italic'})

            # Store handles and labels for the legend
            handles.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10))
            labels.append(species_name)

        ax.set_xlabel('Temperature (deg C)')
        ax.set_ylabel('Pressure (MPa)')
        ax.set_zlabel('Salinity (NaCl %)')

        ax.view_init(elev=view_angle[0], azim=view_angle[1])

        # Create a legend on the middle left side, with additional left offset
        ax.legend(handles, labels, loc='center left', bbox_to_anchor=(0.1, 0.8), bbox_transform=plt.gcf().transFigure, prop={'style': 'italic'})

        plt.show()

species_data = get_species_data()

# List of different view angles (elev, azim)
view_angles = [
    (0, 270),  # Example view angle 1
    (20, 45),  # Example view angle 2
]

plot_species_ovoids(species_data, view_angles)
```

```python
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def plot_ovoid(ax, center, radii, color, alpha=0.2):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    u, v = np.meshgrid(u, v)
    x = radii[0] * np.cos(u) * np.sin(v) + center[0]
    y = radii[1] * np.sin(u) * np.sin(v) + center[1]
    z = radii[2] * np.cos(v) + center[2]

    ax.plot_surface(x, y, z, color=color, alpha=alpha, rstride=1, cstride=1)

def get_species_data():
    species_data = {
        'Dermacoccus abyssi MT1.1T': { 
            'temperature': (10, 37),
            'pressure': (0.1, 40.0),
            'salinity': (7.5, 10.0),
            'optimum': (28, 40.0, 7.5),
        },
        'Carnobacterium sp. strain AT12': {
            'temperature': (2.0, 2,0),
            'pressure': (0.1, 60.0),
            'salinity': (1.95, 1.95),
            'optimum': (2.0, 15.0, 1.95),
        },
        'Carnobacterium sp. strain AT7': {
            'temperature': (20, 20),
            'pressure': (0.1, 60.0),
            'salinity': (1.95, 1.95),
            'optimum': (20, 15.0, 1.95),
        },
        'Anoxybacter fermentans DY22613': {
            'temperature': (44.0, 72.0),
            'pressure': (0.1, 55.0),
            'salinity': (0.65, 5.2),
            'optimum': (61.0, 20.0, 2.3),
        },
        'Piezobacter thermophilus str. 108': {
            'temperature': (30, 55),
            'pressure': (16, 65),
            'salinity': (1.2, 5.4),
            'optimum': (50, 36, 2),
        },
        'Rhodobacterales bacterium PRT1': {
            'temperature': (4, 12),
            'pressure': (10.0, 110),
            'salinity': (3.5, 3.5),
            'optimum': (10, 80, 3.5),
        },
        'Desulfovibrio profundus 500-1': {
            'temperature': (15, 65),
            'pressure': (0.1, 40),
            'salinity': (0.2, 10),
            'optimum': (25, 13, 4.3),
        },
        'Desulfovibrio hydrothermalis AM13': {
            'temperature': (20, 40),
            'pressure': (0.1, 40),
            'salinity': (2.5, 4.0),
            'optimum': (35, 26, 2.5),
        },
        'Desulfovibrio piezophilus C1TLV30': {
            'temperature': (15, 45),
            'pressure': (0.1, 30.0),
            'salinity': (0.1, 8),
            'optimum': (30, 10.0, 2.5),
        },
        'Colwellia marinimaniae MTCD1T': {
            'temperature': (6, 10),
            'pressure': (80, 140),
            'salinity': (3, 3),
            'optimum': (6, 120, 3),
        },
        'Colwellia piezophila Y223G': {
            'temperature': (4, 10),
            'pressure': (40, 80),
            'salinity': (3, 3),
            'optimum': (10, 60, 3),
        },
        'Colwellia piezophila Y251E': {
            'temperature': (10, 10),
            'pressure': (0.1, 80),
            'salinity': (3, 3),
            'optimum': (10, 60, 3),
        },
        'Collwellia hadaliensis BNL1': {
            'temperature': (2, 10),
            'pressure': (37, 102),
            'salinity': (3, 3),
            'optimum': (10, 90, 3),
        },
        'Colwellia sp. strain MT-41': {
            'temperature': (0, 10),
            'pressure': (30, 120),
            'salinity': (3, 3),
            'optimum': (8, 103, 3),
        },
        'Psychromonas profunda 2825': {
            'temperature': (2, 13),
            'pressure': (0.1, 50),
            'salinity': (1.5, 3),
            'optimum': (10, 25, 3),
        },
        'Psychromonas kaikoae JT703': {
            'temperature': (4, 15),
            'pressure': (20, 70),
            'salinity': (1.5, 3),
            'optimum': (10, 50, 3),
        },
        'Psychromonas sp. strain CNPT3': {
            'temperature': (0, 15),
            'pressure': (0.1, 85),
            'salinity': (1.5, 3),
            'optimum': (12, 52, 3),
        },
        'Psychromonas hadalis K41G': {
            'temperature': (6, 10),
            'pressure': (30, 90),
            'salinity': (1.5, 3),
            'optimum': (6, 60, 3),
        },
        'Moritella profunda 2674': {
            'temperature': (2, 12),
            'pressure': (0.1, 50.0),
            'salinity': (1,5, 3),
            'optimum': (6, 25.0, 3),
        },
        'Moritella abyssi 2693': {
            'temperature': (4, 14),
            'pressure': (0.1, 50),
            'salinity': (1.5, 3),
            'optimum': (10, 30, 3),
        },
        'Moritella sp. strain PE36': {
            'temperature': (0, 10),
            'pressure': (0.1, 70),
            'salinity': (3, 3),
            'optimum': (10, 41, 3),
        },
        'Moritella japonica DSK1': {
            'temperature': (10, 15),
            'pressure': (0.1, 70),
            'salinity': (3, 3),
            'optimum': (15, 50, 3),
        },
        'Moritella yayanosii DB21MT-5': {
            'temperature': (10, 10),
            'pressure': (50, 100),
            'salinity': (3, 3),
            'optimum': (10, 80, 3),
        },
        'Shewanella psychrophila WP2': {
            'temperature': (0, 20),
            'pressure': (0.1, 50),
            'salinity': (1, 7.2),
            'optimum': (17, 15, 3.5),
        },
        'Shewanella piezotolerans WP3': {
            'temperature': (0, 28),
            'pressure': (0.1, 50),
            'salinity': (1, 7.2),
            'optimum': (17, 20, 3.5),
        },
        'Shewanella profunda LT13a': {
            'temperature': (4, 37),
            'pressure': (0.1, 50),
            'salinity': (0, 6),
            'optimum': (27, 10, 0.5),
        },
        'Shewanella violacea DSS12': {
            'temperature': (8, 15),
            'pressure': (0.1, 70),
            'salinity': (2, 4),
            'optimum': (10, 30, 3),
        },
        'Shewanella benthica F1A': {
            'temperature': (2, 10),
            'pressure': (0.1, 70),
            'salinity': (3, 3),
            'optimum': (8, 30, 3),
        },
        'Shewanella benthica DB6101': {
            'temperature': (4, 15),
            'pressure': (0.1, 70),
            'salinity': (3, 3),
            'optimum': (10, 50, 3),
        },
        'Shewanella benthica DB5501': {
            'temperature': (4, 15),
            'pressure': (0.1, 70),
            'salinity': (3, 3),
            'optimum': (15, 60, 3),
        },
        'Shewanella benthica DB6705': {
            'temperature': (4, 15),
            'pressure': (20, 70),
            'salinity': (3, 3),
            'optimum': (15, 60, 3),
        },
        'Shewanella benthica DB6906': {
            'temperature': (4, 15),
            'pressure': (20, 70),
            'salinity': (3, 3),
            'optimum': (15, 60, 3),
        },
        'Shewanella benthica DB172R': {
            'temperature': (10, 15),
            'pressure': (20, 70),
            'salinity': (3, 3),
            'optimum': (10, 60, 3),
        },
        'Shewanella benthica DB172F': {
            'temperature': (10, 15),
            'pressure': (50, 80),
            'salinity': (3, 3),
            'optimum': (10, 70, 3),
        },
        'Shewanella benthica DB21MT-2': {
            'temperature': (10, 15),
            'pressure': (60, 100),
            'salinity': (3, 3),
            'optimum': (10, 70, 3),
        },
        'Shewanella benthica KT99': {
            'temperature': (2, 2),
            'pressure': (40, 98),
            'salinity': (3, 3),
            'optimum': (2, 98, 3),
        },
        'Thioprofundum lithotropica str. 106': {
            'temperature': (30, 55),
            'pressure': (0.2, 36),
            'salinity': (1.2, 3),
            'optimum': (50, 16, 4.2),
        },
        'Profundimonas piezophila YC-1': {
            'temperature': (2, 14),
            'pressure': (20, 70),
            'salinity': (1.5, 1.5),
            'optimum': (8, 50, 1.5),
        },
        'Photobacterium profundum DSJ4': {
            'temperature': (4, 10),
            'pressure': (0.1, 70),
            'salinity': (3, 3),
            'optimum': (18, 10, 3),
        },
        'Photobacterium profundum SS9': {
            'temperature': (3, 15),
            'pressure': (37, 102),
            'salinity': (2.3, 2.3),
            'optimum': (15, 28, 2.3),
        },
        'Thermosipho japonica IHB1': {
            'temperature': (45, 80),
            'pressure': (0.1, 60),
            'salinity': (0.3, 6),
            'optimum': (72, 20, 3),
        },
        'Marinitoga piezophila str, KA3': {
            'temperature': (45, 70),
            'pressure': (0.3, 50),
            'salinity': (1, 5),
            'optimum': (10, 65, 3),
        },
        'Methanococcus jannaschii str. JAL-1': {
            'temperature': (50, 90),
            'pressure': (0.1, 75),
            'salinity': (1, 5),
            'optimum': (85, 75, 2.5),
        },
        'Methanococcus thermolithotrophicus': {
            'temperature': (30, 70),
            'pressure': (0.1, 100),
            'salinity': (1.3, 8.3),
            'optimum': (65, 50, 4),
        },
        'Methanopyrus kandleri 116': {
            'temperature': (90, 122),
            'pressure': (0.1, 50),
            'salinity': (0.5, 4.5),
            'optimum': (105, 33, 3),
        },
        'Palaococcus ferrophilus str. DMJ': {
            'temperature': (60, 88),
            'pressure': (0.1, 60),
            'salinity': (2, 7.3),
            'optimum': (83, 60, 4.7),
        },
        'Paleococcus pacificus DY20341': {
            'temperature': (50, 90),
            'pressure': (0.1, 80),
            'salinity': (1, 4),
            'optimum': (80, 30, 3),
        },
        'Pyrococcus abyssi GE5': {
            'temperature': (67, 102),
            'pressure': (0.1, 40),
            'salinity': (1.5, 4.5),
            'optimum': (100, 20, 3),
        },
        'Pyrococcus yayanosii CH1': {
            'temperature': (80, 108),
            'pressure': (20, 120),
            'salinity': (2.5, 5.5),
            'optimum': (98, 52, 3.5),
        },
        'Thermococcus aggregans TY': {
            'temperature': (60, 94),
            'pressure': (0.1, 30),
            'salinity': (1, 3),
            'optimum': (88, 20, 2.5),
        },
        'Thermococcus guaymasensis TYS': {
            'temperature': (56, 90),
            'pressure': (0.1, 50),
            'salinity': (1, 5),
            'optimum': (88, 27.5, 3),
        },
        'Thermococcus peptonophilus OG-1': {
            'temperature': (60, 100),
            'pressure': (0.1, 60),
            'salinity': (1, 5),
            'optimum': (85, 45, 3),
        },
        'Thermococcus eurythermalis A501': {
            'temperature': (50, 102),
            'pressure': (0.1, 70),
            'salinity': (1, 5),
            'optimum': (85, 15, 2.5),
        },
        'Thermococcus barophilus MP': {
            'temperature': (48, 100),
            'pressure': (0.1, 80),
            'salinity': (1, 4),
            'optimum': (85, 40, 2.5),
        },
        'Thermococcus piezophilus CDGS': {
            'temperature': (60, 90),
            'pressure': (0.1, 130),
            'salinity': (2, 6),
            'optimum': (10, 75, 3),
        },
        'Archaeoglobus fulgidus VC-16': {
            'temperature': (60, 95),
            'pressure': (0.1, 60),
            'salinity': (3, 3),
            'optimum': (83, 20, 3),
        }
    }
    return species_data

def plot_species_ovoids(species_data, elevation_angle, azimuth_angle):
    num_species = len(species_data)
    colormap = plt.cm.get_cmap('tab10', num_species)

    fig = plt.figure(figsize=(14, 10))
    ax = fig.add_subplot(111, projection='3d')

    handles = []
    labels = []

    for idx, (species_name, data) in enumerate(species_data.items()):
        center = data['optimum']
        radii = (data['temperature'][1] - data['temperature'][0],
                 data['pressure'][1] - data['pressure'][0],
                 data['salinity'][1] - data['salinity'][0])
        radii = (radii[0] / 2.0, radii[1] / 2.0, radii[2] / 2.0)

        color = colormap(idx / num_species)

        plot_ovoid(ax, center, radii, color=color, alpha=0.2)
        label_offset = 0.5
        ax.text(center[0], center[1], center[2] + label_offset, species_name, fontsize=10, color=color, fontdict={'style': 'italic'})

        handles.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10))
        labels.append(species_name)

    ax.set_xlabel('Temperature (deg C)')
    ax.set_ylabel('pressure')
    ax.set_zlabel('Salinity (NaCl %)')

    # Set initial view angle
    ax.view_init(elev=elevation_angle, azim=azimuth_angle)

    def update(frame):
        ax.view_init(elev=elevation_angle, azim=frame)  # Adjust the view for rotation
        return fig,

    num_frames = 360
    ani = FuncAnimation(fig, update, frames=num_frames, interval=50, blit=True)
    
    ax.legend(handles, labels, loc='center left', bbox_to_anchor=(0.0, 0.8), bbox_transform=plt.gcf().transFigure, prop={'style': 'italic'})
    ani.save('species_rotation.gif', writer='imagemagick', fps=20)
    plt.show()

species_data = get_species_data()
elevation_angle = 20
azimuth_angle = 45

plot_species_ovoids(species_data, elevation_angle, azimuth_angle)
```
